```python
"""
//////////////////////////////////////////////////////////////////
Intento hacer un ejemplo, de una pequeña red neuronal,
a la que suministro datos históricos de los años 19,20,21, y 22.
De medias de población con mediciones de colesterol, trigliceridos,
tensión sanguinea sistólica, y muertes por estas causas.

Luego le suministro unas previsiones poblacionales del Ministerio de Sanidad
que intenta disminuir los baremos de colesterol, trigliceridos, y tension sistolica,
en unos porcentajes determinados el número de muertes, con
campañas de concienciacion y previsión social para los años 23,24, 25, 26.

A la red neuronal deberá, basándose en datos históricos, y las campañas previstas
de concienciación, determinarme gráficamente cuantas muertes habrá en el año
2027. Graficando también los años anteriores.

El número de neuronas de entrada viene especificado como 3,
que corresponde a las características de entrada de
colesterol, triglicéridos y tensión sistólica.
Lleva dos capas de 8 neuronas intermedias, escondidas, y una neurona de
salida

"""
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada y salida históricos en bases de datos
año_2019 = np.array([[234, 190, 13]])  # Datos para el año 2019
año_2020 = np.array([[267, 201, 14]])  # Datos para el año 2020
año_2021 = np.array([[245, 210, 12]])  # Datos para el año 2021
año_2022 = np.array([[264, 234, 15]])  # Datos para el año 2022

muertes_2019 = 123  # Número de muertes en el año 2019
muertes_2020 = 142  # Número de muertes en el año 2020
muertes_2021 = 190  # Número de muertes en el año 2021
muertes_2022 = 210  # Número de muertes en el año 2022

# Construir la red neuronal
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(3,)),  # Capa de entrada
    tf.keras.layers.Dense(8, activation='relu'),                    # Capas intermedias
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1)                                       # Capa de salida con 1 neurona (regresión lineal)
])

# Compilar el modelo
modelo.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo con los datos históricos que conocemos.
años_entrenamiento = np.vstack((año_2019, año_2020, año_2021, año_2022))
muertes_entrenamiento = np.array([[muertes_2019], [muertes_2020], [muertes_2021], [muertes_2022]])

historial_entrenamiento = modelo.fit(años_entrenamiento, muertes_entrenamiento, epochs=1000, batch_size=32, verbose=0)
"""
Realizar predicciones esperadas para los años 2023, 2024, 2025, 2026, 2027, 2028 y 2029,
por aplicaciónes de campañas por radio TV, anuncios en edificios públicos, de conciención
social. Los baremos de colesterol, triglicéridos, y tensión, son los baremos esperados o meta.
 """
año_2023 = np.array([[190, 200, 13]])  # Datos para el año 2023 (completa con datos reales o ficticios)
año_2024 = np.array([[180, 200, 12]])  # Datos para el año 2024 (completa con datos reales o ficticios)
año_2025 = np.array([[170, 220, 12]])  # Datos para el año 2025 (completa con datos reales o ficticios)
año_2026 = np.array([[160, 190, 11]])  # Datos para el año 2026 (completa con datos reales o ficticios)
año_2027 = np.array([[160, 189, 12]])  # Datos para el año 2027 (completa con datos reales o ficticios)
año_2028 = np.array([[170, 178, 11]])  # Datos para el año 2028 (completa con datos reales o ficticios)
año_2029 = np.array([[120, 200, 10]])  # Datos para el año 2029 (completa con datos reales o ficticios)

prediccion_2023 = modelo.predict(año_2023)
prediccion_2024 = modelo.predict(año_2024)
prediccion_2025 = modelo.predict(año_2025)
prediccion_2026 = modelo.predict(año_2026)
prediccion_2027 = modelo.predict(año_2027)
prediccion_2028 = modelo.predict(año_2028)
prediccion_2029 = modelo.predict(año_2029)

# Obtener el muertes para 2023, 2024, 2025, 2026, 2027, 2028 y 2029, con las CAMPAÑAS DE CONCIENCIAÓN SOCIAL.
muertes_2023 = int(np.round(prediccion_2023[0][0]))
muertes_2024 = int(np.round(prediccion_2024[0][0]))
muertes_2025 = int(np.round(prediccion_2025[0][0]))
muertes_2026 = int(np.round(prediccion_2026[0][0]))
muertes_2027 = int(np.round(prediccion_2027[0][0]))
muertes_2028 = int(np.round(prediccion_2028[0][0]))
muertes_2029 = int(np.round(prediccion_2029[0][0]))

print("Número estimado de muertes en 2023:", muertes_2023)
print("Número estimado de muertes en 2024:", muertes_2024)
print("Número estimado de muertes en 2025:", muertes_2025)
print("Número estimado de muertes en 2026:", muertes_2026)
print("Número estimado de muertes en 2027:", muertes_2027)
print("Número estimado de muertes en 2028:", muertes_2028)
print("Número estimado de muertes en 2029:", muertes_2029)

# Graficamos las estimaciones de muertes para los años 2023, 2024, 2025, 2026, 2027, 2028 y 2029
años = ['2022', '2023', '2024', '2025', '2026', '2027', '2028', '2029']
muertes_estimadas = [muertes_2022, muertes_2023, muertes_2024, muertes_2025, muertes_2026, muertes_2027, muertes_2028, muertes_2029]

plt.bar(años, muertes_estimadas, color=['blue', 'green', 'orange', 'purple', 'brown', 'red', 'cyan', 'magenta'])
plt.title('Estimación de Muertes para los años 2023, 2024, 2025, 2026, 2027, 2028 y 2029 aplicando CONCIENCIACIÓN SOCIAL')
plt.xlabel('Año')
plt.ylabel('Número de Muertes ESTIMADAS')
plt.show()

#***************************************************
# Graficar el número de neuronas en cada capa
plt.subplot(2, 1, 1)
capas = ['3 Capa de Entrada', '8 Capa Intermedia', '8 Capa Intermedia', '1 Capa de Salida']
neuronas_por_capa = [3, 8, 8, 1]
plt.bar(capas, neuronas_por_capa, color=['blue', 'green', 'orange', 'purple'])
plt.title('Número de Neuronas en cada Capa')
plt.ylabel('Número de Neuronas')



#*******************************************************

```
