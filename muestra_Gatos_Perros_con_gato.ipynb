```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Directorio de datos
data_dir = '/content/sample_data/'
train_dir = os.path.join(data_dir, 'Cat')  # Directorio de entrenamiento

# Parámetros
input_shape = (32, 32, 3)  # Tamaño de las imágenes reducido
batch_size = 32
epochs = 10

# Obtener lista de imágenes y etiquetas
image_files = [os.path.join(train_dir, file) for file in os.listdir(train_dir) if file.endswith(('jpg', 'jpeg', 'png'))]
images = [img_to_array(load_img(img, target_size=input_shape)) / 255.0 for img in image_files]
labels = [0 if 'gato' in os.path.basename(img).lower() else 1 for img in image_files]

# Convertir a arrays numpy
images = np.array(images)
labels = np.array(labels)

# Definición del modelo
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamiento del modelo
history = model.fit(images, labels, epochs=epochs)

# Carga y procesamiento de la imagen de prueba
sample_image_path = '/content/sample_data/gato.jpg'
sample_image = img_to_array(load_img(sample_image_path, target_size=input_shape)) / 255.0
sample_image_array = np.expand_dims(sample_image, axis=0)

# Predicción de la imagen de prueba
prediction = model.predict(sample_image_array)
class_label = "Gato" if prediction < 0.5 else "Perro"

# Visualización de la imagen de prueba y la predicción
plt.imshow(sample_image[0])
plt.title(f'Predicción: {class_label}')
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Directorio de datos
data_dir = '/content/sample_data/'

# Verificar las imágenes en el directorio 'Cat'
cat_images = os.listdir(os.path.join(data_dir, 'Cat'))
print(f"Number of Cat images: {len(cat_images)}")

# Parámetros
input_shape = (32, 32, 3)
batch_size = 32
epochs = 20

# Cargar manualmente las imágenes de 'Cat'
cat_data = []
for img_name in cat_images:
    img_path = os.path.join(data_dir, 'Cat', img_name)
    img = img_to_array(load_img(img_path, target_size=input_shape[:2])) / 255.0
    cat_data.append(img)

# Etiquetas para las imágenes de gatos
cat_labels = np.zeros(len(cat_data))

# Verificar cuántas imágenes de gatos se cargaron
print(f"Number of loaded Cat images: {len(cat_data)}")

# Convertir a numpy array
cat_data = np.array(cat_data)

# Definición del modelo
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamiento del modelo
history = model.fit(cat_data, cat_labels, epochs=epochs)

# Carga y procesamiento de la imagen de prueba
sample_image_path = '/content/sample_data/gato.jpg'
sample_image = img_to_array(load_img(sample_image_path, target_size=input_shape[:2])) / 255.0
sample_image_array = np.expand_dims(sample_image, axis=0)

# Predicción de la imagen de prueba
prediction = model.predict(sample_image_array)
class_label = "Perro" if prediction >= 0.5 else "Gato"

# Visualización de la imagen de muestra y la predicción
plt.imshow(sample_image[0])
plt.title(f'Predicción: {class_label}')
plt.axis('off')
plt.show()
```
