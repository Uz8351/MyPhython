```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función y su derivada
def f(x):
    return x**2 / 7

def df(x):
    return 2*x / 7

# Parámetros del descenso de gradiente
x_initial = 21
learning_rate = 0.07
iterations = 300

# Listas para almacenar los valores de x y f(x) en cada iteración
x_values = []
f_values = []

# Descenso de gradiente
x_current = x_initial
for i in range(iterations):
    x_values.append(x_current)
    f_values.append(f(x_current))

    gradient = df(x_current)
    x_current -= learning_rate * gradient

# Encontrar el valor mínimo de la derivada
min_gradient_value = min([df(x) for x in x_values])

# Impresión del valor mínimo de la derivada
print("Menor valor de la derivada:", min_gradient_value)

# Gráfica
x_range = np.linspace(-1, 15, 300)

# Graficar la función y el descenso de gradiente
plt.plot(x_range, f(x_range), label='f(x) = x^2 / 7')
plt.scatter(x_values, f(np.array(x_values)), color='black', label='Descenso de Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Ajustar la cuadrícula y las minor ticks en ambos ejes
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='red')
plt.minorticks_on()

plt.tight_layout()

plt.show()
```
