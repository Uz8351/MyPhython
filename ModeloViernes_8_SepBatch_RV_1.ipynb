```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
# Parámetros
learning_rate = 0.01
num_iterations = 100
theta0 = 0
theta1 = 0

# Variables para almacenar la evolución de los coeficientes y el MSE
theta0_history = []
theta1_history = []
mse_history = []

# Descenso de Gradiente
m = len(X)

for i in range(num_iterations):
    # Calcular las predicciones
    y_pred = theta0 + theta1 * X

    # Calcular el error
    error = y_pred - y

    # Actualizar los coeficientes theta0 y theta1 usando el Gradiente Descendente Tipo Batch
    theta0 -= (learning_rate / m) * np.sum(error)
    theta1 -= (learning_rate / m) * np.sum(error * X)

    # Calcular el MSE y almacenarlo en el historial
    mse = np.mean(error**2)
    mse_history.append(mse)

    # Almacenar los coeficientes en el historial
    theta0_history.append(theta0)
    theta1_history.append(theta1)

# Graficar la evolución del MSE con puntos de descenso
plt.figure(figsize=(10, 5))
plt.plot(range(num_iterations), mse_history, label='MSE')
plt.scatter(range(num_iterations), mse_history, c='red', marker='o', label='Descenso', s=50)
plt.xlabel('Iteraciones')
plt.ylabel('MSE')
plt.title('Evolución del MSE durante el Descenso de Gradiente con Puntos de Descenso')
plt.legend()
plt.grid(True)
plt.show()
```
