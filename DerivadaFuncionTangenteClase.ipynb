```python
import numpy as np
import matplotlib.pyplot as plt

def FuncionTangente(x):
    return np.tan(x)

def DerivadaTangente(x):
    return 1 / (np.cos(x) ** 2)

TasaDeAprendizaje = 0.08
Iteraciones = 10000
ValorInicial_x = 6

x_Valores = []
y_Valores = []

x = ValorInicial_x
for _ in range(Iteraciones):
    x_Valores.append(x)
    y_Valores.append(FuncionTangente(x))
    x -= TasaDeAprendizaje * DerivadaTangente(x)

x_Rango = np.linspace(-5, 5, 400)  # Ajusta el rango para una mejor visualización
y_Rango = FuncionTangente(x_Rango)

Minimo_index = np.argmin(y_Valores)
Minimo_x = x_Valores[Minimo_index]
Minimo_y = y_Valores[Minimo_index]

print("Menor valor de x:", Minimo_x)
print("Correspondiente valor de y:", Minimo_y)

plt.figure(figsize=(15, 8))
plt.plot(x_Rango, y_Rango, label='f(x) = tan(x)')
plt.scatter(x_Valores, y_Valores, color='red', label='Puntos de descenso')
plt.scatter(ValorInicial_x, FuncionTangente(ValorInicial_x), color='green', label='Punto inicial')
plt.scatter(Minimo_x, Minimo_y, color='blue', label='Punto final')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente')
plt.xlim(-5, 5)  # Limita el rango de los ejes x para una mejor visualización
plt.ylim(-10, 10)  # Limita el rango de los ejes y para una mejor visualización
plt.grid(True)
plt.legend()
plt.show()
```
