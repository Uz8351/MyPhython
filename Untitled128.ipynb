```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

array_3d = np.random.rand(9, 9, 9)
print("El valor es ->",array_3d[3,4,5])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y, z = np.indices(array_3d.shape)
ax.scatter(x.flatten(), y.flatten(), z.flatten(), c=array_3d.flatten())

# Agregar etiqueta al punto correspondiente
ax.text(3, 4, 5, array_3d[3, 4, 5], color='black')

plt.show()

```
