```python
import numpy as np
import matplotlib.pyplot as plt

t = 3.14

def my_funct(x):
    return (x ** 2 / 2) + t

def derivada(x):
    return x

x = np.linspace(-20, 20, 200)
y = my_funct(x)

init_x = 15  # Punto de inicio para el descenso del gradiente
lr = 0.05  # La tasa de aprendizaje (learning rate en inglés)
init_y = my_funct(init_x)

px = init_x - derivada(init_x) * lr
py = my_funct(px)

plt.plot(x, y)
plt.plot(init_x, init_y, 'bo')
plt.plot(px, py, 'bo')

min_derivative = derivada(px)  # Inicializar con el valor de la primera derivada
min_px = px

for i in range(5000):
    px = px - derivada(px) * lr
    py = my_funct(px)
    plt.plot(px, py, 'bo')

    current_derivative = derivada(px)
    if current_derivative < min_derivative:
        min_derivative = current_derivative
        min_px = px

plt.show()

print("Valor de px en el punto de la derivada más pequeña:", format(min_px, '.125f'))
```
