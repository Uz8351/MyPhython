```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de costo
def costo(x):
    return (x**2 + 6) / 12

# Definir el gradiente de la función de costo
def gradiente(x):
    return (x / 6)

# Parámetros iniciales
x = 18
learning_rate = 0.1
num_iteraciones = 100

# Listas para almacenar los valores de "x" y el costo en cada iteración
x_historial = []
costo_historial = []

# Descenso de gradiente
for i in range(num_iteraciones):
    x_historial.append(x)
    costo_actual = costo(x)
    costo_historial.append(costo_actual)
    grad = gradiente(x)
    x = x - learning_rate * grad

# Graficar la función de costo
x_valores = np.linspace(0, 30, 100)
y_valores = [(v**2 + 6) / 12 for v in x_valores]

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.plot(x_valores, y_valores, label='Función de Costo')
plt.scatter(x_historial, costo_historial, c='r', label='Descenso de Gradiente')
plt.xlabel('x')
plt.ylabel('Costo')
plt.legend()

# Graficar los valores de costo MSE
plt.subplot(122)
plt.plot(range(num_iteraciones), costo_historial, marker='o', linestyle='-', color='b')
plt.xlabel('Iteraciones')
plt.ylabel('Costo MSE')
plt.title('Valores de Costo MSE en Iteraciones')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de costo
def costo(x):
    return (x**2 + 6) / 12

# Definir el gradiente de la función de costo
def gradiente(x_batch):
    return (x_batch / 6)

# Parámetros iniciales
x = 18
learning_rate = 0.1
num_iteraciones = 100
batch_size = 10  # Tamaño del lote (batch size)

# Listas para almacenar los valores de "x" y el costo en cada iteración
x_historial = []
costo_historial = []

# Datos de ejemplo para un conjunto de datos ficticio
datos = np.linspace(1, 30, 100)  # Simulación de un conjunto de datos

# Descenso de gradiente tipo Batch
for i in range(num_iteraciones):
    x_historial.append(x)
    costo_actual = costo(x)
    costo_historial.append(costo_actual)

    # Seleccionar un lote de datos
    batch_indices = np.random.choice(len(datos), batch_size, replace=False)
    x_batch = datos[batch_indices]

    # Calcular el gradiente utilizando el lote de datos
    grad = gradiente(x_batch.mean())

    # Actualizar el valor de "x"
    x = x - learning_rate * grad

# Graficar la función de costo
x_valores = np.linspace(0, 30, 100)
y_valores = [(v**2 + 6) / 12 for v in x_valores]

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.plot(x_valores, y_valores, label='Función de Costo')
plt.scatter(x_historial, costo_historial, c='r', label='Descenso de Gradiente')
plt.xlabel('x')
plt.ylabel('Costo')
plt.legend()

# Graficar los valores de costo MSE
plt.subplot(122)
plt.plot(range(num_iteraciones), costo_historial, marker='o', linestyle='-', color='b')
plt.xlabel('Iteraciones')
plt.ylabel('Costo MSE')
plt.title('Valores de Costo MSE en Iteraciones')
plt.grid(True)

plt.tight_layout()
plt.show()
```
