```python
"""
Función tangente hiperbólica:
****************************
Los valores de salida de la función
tangente hiperbólica están comprendidos dentro de
un rango que va de -1 a 1. Al modificar el valor de
de uno de sus parametros
se ve afectada la pendiente de la función de
activación.

La fórmula matemática de la tangente hiperbólica (tanh) es la siguiente:

tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

Donde:

x es el valor de entrada.
e es la base del logaritmo natural (aproximadamente 2.71828).
La función tangente hiperbólica mapea cualquier valor real en el rango [-1, 1].
A medida que x se acerca a infinito positivo, tanh(x) se aproxima a 1.
Mientras que cuando x se acerca a infinito negativo, tanh(x) se aproxima a -1.

Esta fórmula es la base para calcular la tangente hiperbólica en
diversas aplicaciones y algoritmos de redes neuronales.


"""
#Importamos los modulos necesarios para este ejercicio.
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def update_plot(a=1, b=1):
    x = np.linspace(-10, 10, 1000)
    y = a * np.tanh(b * x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Tangente Hiperbólica', color='blue')
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.xlabel('x')
    plt.ylabel('tanh(ax)')
    plt.title('Función Tangente Hiperbólica')
    plt.legend()
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(0.5)
    plt.gca().spines['bottom'].set_linewidth(0.5)

    plt.show()

# Creamos los controles deslizante para jugar con la curva los controles deslizantes
a_slider = widgets.FloatSlider(value=1, min=0.1, max=10, step=0.1, description='a:')
b_slider = widgets.FloatSlider(value=1, min=0.1, max=10, step=0.1, description='b:')

# Llamar a la función de actualización cuando se modifiquen los controles deslizantes
widgets.interact(update_plot, a=a_slider, b=b_slider)

# Mostrar los controles deslizantes
display(a_slider, b_slider)
```
