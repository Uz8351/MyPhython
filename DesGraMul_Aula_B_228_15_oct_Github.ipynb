```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función f(x, y)
def f(x, y):
    return 7 * x * y / np.exp(x**2 + y**2)

# Gradiente de la función f(x, y)
def gradient(x, y):
    df_dx = (7 * y * (1 - 2 * x**2) * np.exp(-(x**2 + y**2))) / (x**2 + y**2)
    df_dy = (7 * x * (1 - 2 * y**2) * np.exp(-(x**2 + y**2))) / (x**2 + y**2)
    return df_dx, df_dy

# Inicialización de parámetros
learning_rate = 0.1
num_iterations = 100
x = 1.03
y = 1.78
losses = []

# Realizar el descenso de gradiente
for i in range(num_iterations):
    grad_x, grad_y = gradient(x, y)
    x -= learning_rate * grad_x
    y -= learning_rate * grad_y
    loss = f(x, y)
    losses.append(loss)

# Crear una malla de puntos para la gráfica 3D
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Crear las figuras
fig = plt.figure(figsize=(14, 8))
fig.suptitle("Descenso de Gradiente Múltiple para f(x, y)")

# Gráfica 3D de la función
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')

# Gráfica del descenso de gradiente lineal múltiple
ax2 = fig.add_subplot(132)
ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.plot(x, y, 'ro-')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

# Gráfica de la pérdida MSE
ax3 = fig.add_subplot(133)
ax3.plot(range(num_iterations), losses)
ax3.set_xlabel('Iteraciones')
ax3.set_ylabel('Pérdida (MSE)')

plt.show()
```
