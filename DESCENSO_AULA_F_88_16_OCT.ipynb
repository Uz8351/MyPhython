```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x, y)
def f(x, y):
    return (7 * x * y) / 3**(x**2 + y**2)

# Crear una malla de puntos para graficar las líneas de nivel
x_range = np.linspace(-2, 2, 400)
y_range = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Crear la figura para las líneas de nivel en planta
plt.figure(figsize=(10, 3))


# Crear listas para almacenar los valores de descenso de gradiente
x_values = [1]
y_values = [4]
mse_values = [0.5]  # Valor inicial

# Gradiente de la función f(x, y)
def gradient(x, y):
    df_dx = (7 * y * np.log(3) * 3**(x**2 + y**2) - 7 * x**2 * y * np.log(3) * 3**(x**2 + y**2)) / (3**(x**2 + y**2))**2
    df_dy = (7 * x * np.log(3) * 3**(x**2 + y**2) - 7 * x * y**2 * np.log(3) * 3**(x**2 + y**2)) / (3**(x**2 + y**2))**2
    return df_dx, df_dy

# Tasa de aprendizaje
learning_rate = 0.1

# Número de iteraciones
num_iterations = 20

for _ in range(num_iterations):
    x, y = x_values[-1], y_values[-1]
    df_dx, df_dy = gradient(x, y)
    x_new = x - learning_rate * df_dx
    y_new = y - learning_rate * df_dy
    x_values.append(x_new)
    y_values.append(y_new)
    mse_values.append(f(x_new, y_new))

# Gráfica del descenso de gradiente MSE
plt.figure(figsize=(6, 5))
plt.plot(range(len(mse_values)), mse_values, marker='o', linestyle='-')
plt.xlabel('Iteración')
plt.ylabel('MSE')
plt.title('Descenso de Gradiente MSE')

# Mostrar las gráficas
plt.tight_layout()
plt.show()
# Crear la figura en 3D
fig =plt.figure(figsize=(14, 7))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
# Graficar las líneas de nivel con valores etiquetados
x_range = np.linspace(-2, 2, 400)
y_range = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Crear la figura para las líneas de nivel en planta
plt.figure(figsize=(12, 5))

contour = plt.contour(X, Y, Z, colors='gray', levels=np.arange(0, 1, 0.05))
plt.clabel(contour, inline=1, fontsize=10)
plt.show()
```
