```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 2 * X2 + np.random.randn(100, 1)

# Agregar una columna de unos para el trmino de sesgo (intercepto)
X_b = np.c_[np.ones((100, 1)), X1, X2]

# Inicializar parmetros theta manualmente (coeficientes)
theta = np.random.randn(3, 1)

# Hiperparmetros del descenso de gradiente
tasa_aprendizaje = 0.1
iteraciones = 1000
m = len(y)

# Funcin de costo MSE
def calcular_costo(theta, X, y):
    predicciones = X.dot(theta)
    error = predicciones - y
    costo = (1 / (2 * m)) * np.sum(error**2)
    return costo

# Descenso de gradiente
costos = []
for _ in range(iteraciones):
    gradiente = (1 / m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= tasa_aprendizaje * gradiente
    costo = calcular_costo(theta, X_b, y)
    costos.append(costo)

# Graficar datos de entrada X1 y X2 y la lnea de regresin
plt.figure(figsize=(12, 6))
ax = plt.axes(projection='3d')
ax.scatter3D(X1, X2, y, c='b', label='Datos de Entrada')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Datos de Entrada y Regresin Lineal')

X1_plot, X2_plot = np.meshgrid(X1, X2)
y_plot = theta[0][0] + theta[1][0] * X1_plot + theta[2][0] * X2_plot
ax.plot_surface(X1_plot, X2_plot, y_plot, cmap='viridis', alpha=0.7, label='Lnea de Regresin')

# Graficar la funcin de costo a lo largo de las iteraciones
plt.figure(figsize=(6, 6))
plt.plot(range(iteraciones), costos)
plt.xlabel("Iteracin")
plt.ylabel("Costo MSE")
plt.title("Descenso de Gradiente - Costo MSE")

plt.show()

# Imprimir los parmetros finales del modelo
print("Parmetros finales del modelo:")
print("theta0:", theta[0][0])
print("theta1:", theta[1][0])
print("theta2:", theta[2][0])
```
