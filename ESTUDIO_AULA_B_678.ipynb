```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funcin cuadrtica de ejemplo
def quadratic_function(x):
    return 2 * x**2 + 3 * x + 1

# Definir el gradiente de la funcin
def gradient(x):
    return 4 * x + 3

# Parmetros del descenso de gradiente
learning_rate = 0.1
num_epochs = 10
batch_size = 4  # Tamao del lote

# Inicializar el valor inicial de x
x = 5

# Listas para almacenar valores de x e historial de prdida
x_history = [x]
loss_history = []

# Descenso de gradiente
for epoch in range(num_epochs):
    for _ in range(0, 20, batch_size):  # Usamos un rango fijo de 20 para este ejemplo
        # Calcular el gradiente en el lote actual
        grad = 0
        for i in range(batch_size):
            grad += gradient(x)

        grad /= batch_size

        # Actualizar x usando el descenso de gradiente
        x -= learning_rate * grad

        # Calcular la prdida y guardar en el historial
        loss = quadratic_function(x)
        loss_history.append(loss)

    x_history.append(x)

# Crear un rango de valores de x para graficar la parbola
x_range = np.linspace(-10, 10, 400)
y_range = quadratic_function(x_range)

# Visualizar el proceso de descenso de gradiente
plt.figure(figsize=(12,6))
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_history,color='black',marker='o',markerfacecolor='red', linestyle='-')
plt.title("Optimizacin de x durante el descenso de gradiente")
plt.xlabel("poca")
plt.ylabel("Valor de x")

plt.subplot(1, 2, 2)
plt.plot(loss_history, color='black',marker='o',markerfacecolor='red', linestyle='-')
plt.title("Prdida durante el descenso de gradiente")
plt.xlabel("Iteracin")
plt.ylabel("Valor de prdida")

plt.tight_layout()
plt.show()

# Visualizar el proceso de descenso de gradiente sobre la grfica de la parbola
plt.figure(figsize=(14, 6))
plt.plot(x_range, y_range, label='2 * x^2 + 3 * x + 1 (Funcin de prdida)', linestyle='--')
plt.plot(x_history, [quadratic_function(xi) for xi in x_history], marker='o', linestyle='-', label='Descenso de gradiente')
plt.title("Descenso de Gradiente sobre la Grfica de la Parbola")
plt.xlabel("Valor de x")
plt.ylabel("Valor de y")
plt.legend()
plt.grid(True)
plt.show()
```
