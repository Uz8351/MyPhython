```python
import numpy as np
import matplotlib.pyplot as plt

# Función de costo
def cost_function(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    error = (predictions - y)
    cost = (1/(2*m)) * np.dot(error.T, error)
    return cost

# Gradiente
def gradient(X, y, theta):
    m = len(y)

    predictions = X.dot(theta)
    error = (predictions - y)
    grad = (1/m) * X.T.dot(error)
    return grad

# Datos (X como matriz bidimensional)
X = np.array([[1, 1], [1, 3], [1, 5], [1, 3], [1, 7], [1, 15],
          [1, 11], [1, 13], [1, 15], [1, 13], [1, 17], [1, 15],
          [1, 1], [1, 13], [1, 5], [1, 13], [1, 17], [1, 15], [1, 11],
          [1, 13], [1, 15], [1, 13], [1, 17], [1, 15]])

y = np.array([2.33, 7.0, 14.33, 5.0, 20.33, 83.0, 42.33, 60.33, 81.0, 58.33,
              100.33, 83.0, 12.33, 60.33, 24.33, 58.33, 100.33, 93.0, 52.33,
              70.33, 91.0, 68.33, 110.33, 93.0])

# Theta inicial
theta = np.array([0, 0])  # Theta es un array unidimensional

# Hiperparámetros
learning_rate = 0.01
iterations = 300

# Historial de costo
cost_history = np.zeros(iterations)

# Historial de theta para la línea de descenso
theta_history = []

# Descendiendo gradiente
for i in range(iterations):
    cost = cost_function(X, y, theta)
    cost_history[i] = cost
    grad = gradient(X, y, theta)
    theta = theta - (learning_rate * grad)
    theta_history.append(theta.copy())  # Guardar el valor de theta en esta iteración

# Puntos para la línea de regresión
x_line = np.linspace(min(X[:, 1]), max(X[:, 1]), 120)
y_line = theta[0] + theta[1] * x_line

# Ajustar el tamaño de la figura
plt.figure(figsize=(10, 6))  # Ancho x Alto

# Gráfica
plt.scatter(X[:, 1], y, label='Datos de entrenamiento')
plt.plot(x_line, y_line, color='red', label='Línea de regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión lineal con descenso de gradiente batch')
plt.legend()
plt.grid(True)
plt.show()
```
