```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de muestra proporcionados
X = np.array([[1, 1], [1, 3], [1, 5], [1, 3], [1, 7], [1, 15],
              [1, 11], [1, 13], [1, 15], [1, 13], [1, 17], [1, 15],
              [1, 1], [1, 13], [1, 5], [1, 13], [1, 17], [1, 15], [1, 11],
              [1, 13], [1, 15], [1, 13], [1, 17], [1, 15]])

y = np.array([1.33, 7.0, 14.33, 5.0, 20.33, 83.0, 42.33, 60.33, 81.0, 57.33,
              100.33, 84.90, 12.33, 63.33, 24.33, 59.33, 102.33, 93.0, 52.33,
              70.33, 91.0, 68.33, 110.33, 96.90])

# Inicializar parámetros m y b
m = 0
b = 0

# Declaramos los parámetros del descenso de gradiente
learning_rate = 0.01
iterations = 100

# Listas para almacenar el historial de MSE
mse_history = []

# Descenso de gradiente tipo Batch
for i in range(iterations):
    # Calcular las predicciones del modelo
    y_pred = np.dot(X, np.array([b, m]))

    # Calcular el error cuadrático medio (MSE)
    mse = np.mean((y_pred - y) ** 2)
    mse_history.append(mse)

    # Calcular el gradiente de la función de costo (MSE)
    gradient = np.mean((y_pred - y)[:, np.newaxis] * X, axis=0)

    # Actualizar los parámetros m y b
    b = b - learning_rate * gradient[0]
    m = m - learning_rate * gradient[1]

# Hacer predicciones en nuevos datos
new_x = np.array([1, 1.5])  # Nuevo valor de x
new_y = np.dot(new_x, np.array([b, m]))

# Graficar los datos y la línea de regresión
plt.scatter(X[:, 1], y, label='Datos de muestra')
plt.plot(X[:, 1], np.dot(X, np.array([b, m])), color='red', label='Línea de regresión')
plt.scatter(new_x[1], new_y, color='green', marker='o', label='Predicción')
plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.title('Regresión lineal con descenso de gradiente')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir los parámetros finales de la regresión
print("Parámetros finales:")
print(f"m = {m}")
print(f"b = {b}")
```
