```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos una funcin de costo (por ejemplo, una funcin cuadrtica)
def cost_function(x):
    return x**2+9*x + 6

# Gradiente de la funcin de costo
def gradient(x):
    return 2*x + 9

# Parmetros del algoritmo
learning_rate = 0.1
num_iterations = 30
momentum = 0.9  # Factor de momentum

# Inicializacin de variables
x = 8  # Valor inicial
velocity = 0  # Inicializacin del momentum

# Listas para almacenar la historia de valores y costos
x_history = []
cost_history = []


for _ in range(num_iterations):
    # Calcular el gradiente con respecto al valor estimado (Nesterov)
    gradient_est = gradient(x + momentum * velocity)

    # Actualizar el momentum
    velocity = momentum * velocity - learning_rate * gradient_est

    # Actualizar el valor estimado
    x = x + velocity

    # Calcular el costo y almacenar en las listas de historia
    cost = cost_function(x)
    x_history.append(x)
    cost_history.append(cost)
for index, miscostos in enumerate(cost_history):
    print(f'Iteracions: {index} Costos{miscostos}')
# Graficar el historial de descenso
plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.plot(x_history, cost_history, linestyle='-',color='black',marker='o',markerfacecolor='red')
plt.title("Descenso de Gradiente de Nesterov")
plt.xlabel("Iteracin")
plt.ylabel("Costo")

plt.subplot(1, 2, 2)
x_values = np.linspace(-5, 2, 100)
y_values = cost_function(x_values)
plt.plot(x_values, y_values, linestyle='--', color='gray')
plt.plot(x_history, cost_history, marker='o', linestyle='-')
plt.title("Descenso de Gradiente de Nesterov - Trayectoria")
plt.xlabel("x")
plt.ylabel("Costo")

plt.tight_layout()
plt.show()
```
