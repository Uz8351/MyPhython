```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = 2 * np.random.rand(10, 1)
y = 4 + 3 * X + np.random.rand(10, 1)
print("X:\n", X)
print("******************")
print("y:\n", y)
# Agregar el término de sesgo a X
X_b = np.c_[np.ones((10, 1)), X]
print("****************")
print("X_b:\n",X_b)
# Inicializar parámetros y hiperparámetros
learning_rate = 0.1
n_iterations = 100
m = len(X)  # número de ejemplos de entrenamiento
np.random.seed(1)
# Inicializar parámetros del modelo
theta = np.random.rand(2, 1)
print("***********************")
print("theta:\n",theta)
# Almacenar la evolución del error cuadrático medio (MSE)
mse_history = []

# Descenso de gradiente tipo Batch
for iteration in range(n_iterations):
    gradients = -2/m * X_b.T.dot(y - X_b.dot(theta))
    theta -= learning_rate * gradients
    mse = np.mean((X_b.dot(theta) - y)**2)
    mse_history.append(mse)

# Plotear la evolución del MSE con puntos rojos en los saltos
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(range(n_iterations), mse_history, marker='o', markerfacecolor='red', markersize=5)
plt.xlabel('Iteraciones')
plt.ylabel('Error Cuadrático Medio (MSE)')
plt.title('Evolución del Error Cuadrático Medio (MSE) con Saltos')

# Plotear el ajuste lineal
plt.subplot(1, 2, 2)
plt.scatter(X, y, label='Datos de entrenamiento')
plt.plot(X, X_b.dot(theta), color='red', label='Ajuste Lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Ajuste Lineal con Descenso de Gradiente')
plt.legend()
plt.show()

print("Parámetros finales del modelo:")
print("Theta_0:", theta[0][0])
print("Theta_1:", theta[1][0])
```
