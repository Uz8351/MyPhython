```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de muestra sintticos
np.random.seed(0)
X = 2 * np.random.rand(102, 1)
y = 4 + 3 * X + np.random.randn(102, 1)
print("X:\n",X)
print("*****************")
print("y:\n",y)
print("*****************")
# Inicializar parmetros m y b
m = 0
b = 0

# Hiperparmetros del descenso de gradiente
learning_rate = 0.8
iterations = 200

# Listas para guardar los valores de prdida y parmetros en cada iteracin
losses = []
ms = []
bs = []

# Descenso de gradiente tipo Batch
for i in range(iterations):
    # Calcular las predicciones del modelo
    y_pred = m * X + b

    # Clcular el gradiente de la funcin de costo (MSE)
    gradient_m = (2/iterations) * np.mean((y_pred - y) * X)
    gradient_b = (2/iterations) * np.mean((y_pred - y) * X)

    # Actualizar los parmetros m y b
    m = m - learning_rate * gradient_m
    b = b - learning_rate * gradient_b

    # Calcular y almacenar la prdida en esta iteracin
    loss = np.mean((y_pred - y) ** 2)
    losses.append(loss)
    ms.append(m)
    bs.append(b)

print("y_pred:\n:",y_pred)
print("************************")
#---------------------------------------------------------
# Asumiendo que las tres listas tienen la misma longitud
for i in range(len(losses)):
    print(f"iterations {i + 1}: losses={losses[i]}, ms={ms[i]}, bs={bs[i]}")



#-------------------------------------------------------




# Hacer predicciones en nuevos datos
new_x = np.array([[1.5]])  # Nuevo valor de x
new_y = m * new_x + b

# Graficar la prdida en funcin de las iteraciones
plt.figure(figsize=(15, 7))
plt.subplot(1, 2, 1)
plt.plot(range(iterations), losses, label='Prdida')
plt.scatter(range(iterations), losses, c='red', marker='o', label='Valores de Prdida')
plt.xlabel('Iteraciones')
plt.ylabel('Prdida')
plt.title('Prdida en funcin de las Iteraciones')
plt.legend()

# Graficar los datos y la lnea de regresin
plt.subplot(1, 2, 2)
plt.scatter(X, y, label='Datos de muestra')
plt.plot(X, m * X + b, color='red', label='Lnea de regresin')
plt.scatter(new_x, new_y, color='green', marker='o', label='Prediccin')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresin lineal con descenso de gradiente')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Imprimir los parmetros finales del modelo
print(f"Parmetro m: {m}")
print(f"Parmetro b: {b}")
```

```python
import numpy as np

# Datos de entrenamiento
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Parmetros iniciales del modelo
theta0 = 0.5  # Pendiente
theta1 = 1.0  # Trmino de sesgo

# Predicciones del modelo
predictions = theta0 * X + theta1

# Clculo del gradiente de MSE
n = len(X)
gradient_theta0 = (2/n) * np.sum((predictions - y) * X)
gradient_theta1 = (2/n) * np.sum(predictions - y)

# Actualizacin de los parmetros utilizando el descenso de gradiente
learning_rate = 0.08
theta0 = theta0 - learning_rate * gradient_theta0
theta1 = theta1 - learning_rate * gradient_theta1

# Los nuevos valores de theta0 y theta1 despus de una iteracin
print("Nuevo valor de theta0:", theta0)
print("Nuevo valor de theta1:", theta1)
```
