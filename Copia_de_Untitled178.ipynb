```python
import numpy as np

# Función ReLU
def relu(x):
    return np.maximum(0, x)

# Derivada de la función ReLU
def relu_derivada(x):
    return np.where(x > 0, 1, 0)

# Datos de entrenamiento
X = np.array([[50, 2],     # Presión sistólica  1:  2 Tomas de tensión.
              [70, 3],     # Presión sistólica  1:  3 Tomas de tensión.
              [90, 5],     # Presión sistólica  1:  5 Tomas de tensión.
              [90, 4],     # Presión sistólica  1:  4 Tomas de tensión.
              [80, 6],     # Presión sistólica  1:  6 Tomas de tensión.
              [80, 6]])    # Presión sistólica  1:  6 Tomas de tensión.

# Valores objetivo (ajustados para que coincida con el número de Tomas)
y = np.array([[0.5], [0.7], [0.9], [0.8], [0.6], [0.85]])

# Pesos de la capa oculta (6 neuronas, 2 entradas) - Inicializados aleatoriamente y divididos por 100
W1 = np.random.randn(2, 6) / 100

# Pesos de la capa de salida (6 neuronas en oculta, 1 neurona en salida) - Inicializados aleatoriamente y divididos por 100
W2 = np.random.randn(6, 1) / 100

# Tasa de aprendizaje
learning_rate = 0.001

# Entrenamiento durante 1000 épocas
for epoch in range(100):
    # Propagación hacia adelante
    capa_oculta = np.dot(X, W1)
    capa_oculta_act = relu(capa_oculta)
    salida = np.dot(capa_oculta_act, W2)
    salida_act = relu(salida)

    # Calcular el error
    error = y - salida_act

    # Backpropagation
    d_salida = error * relu_derivada(salida)
    d_W2 = np.dot(capa_oculta_act.T, d_salida) / X.shape[0]
    d_W1 = np.dot(X.T, np.dot(d_salida, W2.T) * relu_derivada(capa_oculta)) / X.shape[0]

    # Actualizar pesos y sesgos
    W2 += learning_rate * d_W2
    W1 += learning_rate * d_W1

    # Mostrar el error en cada época
    mse = np.mean(np.square(error))
    print(f"Época {epoch + 1}/{1000}, Epoca, iteración completa-Error---->: {mse}")

# Imprimir la salida después de 1000 épocas de entrenamiento
print("Salida después de entrenamiento:")
print(salida_act)
```
