```python
#Ejercio modificado de código fuente inicial:
#https://www.aprendemachinelearning.com/una-sencilla-red-neuronal-en-python-con-keras-y-tensorflow/
#Se incluyen graficas
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
import matplotlib.pyplot as plt

# Cargamos las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# Resultados esperados para las combinaciones de la compuerta XOR
target_data = np.array([[0],[1],[1],[0]], "float32")

model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

# Creamos listas vacías para almacenar los valores de pérdida y precisión
loss_history = []
accuracy_history = []

# Entrenamos el modelo y almacenamos los valores de pérdida y precisión en cada época
for epoch in range(1000):
    history = model.fit(training_data, target_data, epochs=1, verbose=0)
    loss = history.history['loss'][0]
    accuracy = history.history['binary_accuracy'][0]
    loss_history.append(loss)
    accuracy_history.append(accuracy)

    # Imprimimos el progreso cada 100 épocas
    if (epoch + 1) % 100 == 0:
        print("Epoch:", epoch+1, "Loss:", loss, "Accuracy:", accuracy)

# Evaluamos el modelo
scores = model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# Calculamos las predicciones de la red neuronal
predictions = model.predict(training_data).round()

# Calculamos el grado de aprendizaje
accuracy = np.mean(predictions == target_data) * 100
print("Grado de aprendizaje: %.2f%%" % accuracy)

# Graficamos la pérdida y precisión a lo largo de las épocas
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(loss_history)
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Vueltas')
plt.ylabel('Pérdida')

plt.subplot(1, 2, 2)
plt.plot(accuracy_history)
plt.title('Precisión durante el entrenamiento')
plt.xlabel('Vueltas')
plt.ylabel('Precisión')

plt.tight_layout()
plt.show()
```
