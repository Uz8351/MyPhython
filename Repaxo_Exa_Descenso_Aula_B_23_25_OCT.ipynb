```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def f(x):
    return (x + 3)**2 / 18

# Gradiente de la función objetivo
def gradient(x):
    return (x + 3) / 9

# Parámetros del descenso de gradiente tipo Batch
alpha = 0.1
iterations = 100
x = 28
cost_history = []

# Valores a lo largo del rango de la función
x_values = np.linspace(-10, 40, 100)

# Realizar descenso de gradiente tipo Batch
for i in range(iterations):
    gradient_sum = 0
    for x_j in x_values:
        gradient_sum += gradient(x_j)
    gradient_value = gradient_sum / len(x_values)  # Promedio del gradiente
    x = x - alpha * gradient_value
    cost = ((f(x) - f(0))**2) / 2  # Costo MSE
    cost_history.append(cost)

# Gráfica de la función
y_values = f(x_values)
plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, label='Función f(x)')
plt.scatter(x, f(x), color='red', label='Puntos de descenso', marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente Tipo Batch - Función f(x)')
plt.legend()
plt.grid(True)

# Gráfica del costo MSE
plt.figure(figsize=(12, 6))
plt.plot(range(1, iterations + 1), cost_history, marker='o', color='b', label='Costo MSE')
plt.xlabel('Iteraciones')
plt.ylabel('Costo MSE')
plt.title('Evolución del Costo MSE durante el Descenso de Gradiente Tipo Batch')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def f(x):
    return (x + 3)**2 / 18

# Gradiente de la función objetivo
def gradient(x):
    return (x + 3) / 9

# Parámetros del descenso de gradiente con momentum
alpha = 0.1
beta = 0.9  # Factor de momentum
iterations = 10
x = 28
v = 0  # Inicialización del momentum
cost_history = []

# Realizar descenso de gradiente con momentum
for i in range(iterations):
    gradient_value = gradient(x)
    v = beta * v + (1 - beta) * gradient_value
    x = x - alpha * v
    cost = ((f(x) - f(0))**2) / 2  # Costo MSE
    cost_history.append(cost)

# Gráfica de la función
x_values = np.linspace(-10, 40, 100)
y_values = f(x_values)
plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, label='Función f(x)')
plt.scatter(x, f(x), color='red', label='Puntos de descenso', marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente con Momentum - Función f(x)')
plt.legend()
plt.grid(True)

# Gráfica del costo MSE
plt.figure(figsize=(12, 6))
plt.plot(range(1, iterations + 1), cost_history, marker='o', color='b', label='Costo MSE')
plt.xlabel('Iteraciones')
plt.ylabel('Costo MSE')
plt.title('Evolución del Costo MSE durante el Descenso de Gradiente con Momentum')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def f(x):
    return (x + 3)**2 / 18

# Gradiente de la función objetivo
def gradient(x):
    return (x + 3) / 9

# Parámetros del descenso de gradiente con momentum
alpha = 0.1
beta = 0.9  # Factor de momentum
iterations = 10
x = 28
v = 0  # Inicialización del momentum
cost_history = []
x_history = []  # Lista para almacenar los puntos del descenso

# Realizar descenso de gradiente con momentum
for i in range(iterations):
    gradient_value = gradient(x)
    v = beta * v + (1 - beta) * gradient_value
    x = x - alpha * v
    cost = ((f(x) - f(0))**2) / 2  # Costo MSE
    cost_history.append(cost)
    x_history.append(x)

# Gráfica de la función y puntos del descenso
x_values = np.linspace(-10, 40, 100)
y_values = f(x_values)
plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, label='Función f(x)')
plt.scatter(x_history, [f(x) for x in x_history], color='red', label='Puntos de descenso', marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente con Momentum - Función f(x) y Puntos de Descenso')
plt.legend()
plt.grid(True)

# Gráfica del costo MSE
plt.figure(figsize=(12, 6))
plt.plot(range(1, iterations + 1), cost_history, marker='o', color='b', label='Costo MSE')
plt.xlabel('Iteraciones')
plt.ylabel('Costo MSE')
plt.title('Evolución del Costo MSE durante el Descenso de Gradiente con Momentum')
plt.grid(True)

plt.show()
```
