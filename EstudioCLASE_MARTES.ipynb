```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def function(x):
    return 2*x + 5*x/3 + x**2

def derivative(x):
    return 2 + 5/3 + 2*x

# Parámetros iniciales
x_start = 18
learning_rate = 0.02
iterations = 320

# Inicializar listas para almacenar valores
x_values = [x_start]
y_values = [function(x_start)]

# Descenso de gradiente
for _ in range(iterations):
    x_new = x_values[-1] - learning_rate * derivative(x_values[-1])
    x_values.append(x_new)
    y_values.append(function(x_new))

# Graficar la función y el descenso de gradiente
x_range = np.linspace(-20, 20, 400)
y_range = function(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Función: $2x + \\frac{5x}{3} + x^2$')
plt.scatter(x_values, y_values, color='red', label='Descenso de Gradiente')
plt.scatter(x_start, function(x_start), color='green', label='Inicio (x=18)')
plt.scatter(x_values[-1], y_values[-1], color='blue', label='Mínimo cercano a 0')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Descenso de Gradiente para $2x + \\frac{5x}{3} + x^2$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
