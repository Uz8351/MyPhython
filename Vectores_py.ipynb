```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def representar_vector(vector, color, label):
    # Desempaquetar las componentes del vector
    x, y, z = vector

    # Crear la figura y el gráfico 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Representar el vector como una flecha en el origen (0, 0, 0)
    ax.quiver(0, 0, 0, x, y, z, color=color, label=label, linewidth=2)

    # Configurar los ejes
    max_coordinate = max(x, y, z)
    ax.set_xlim([0, max_coordinate + 1])
    ax.set_ylim([0, max_coordinate + 1])
    ax.set_zlim([0, max_coordinate + 1])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Agregar una etiqueta al final del vector con las coordenadas
    ax.text(x, y, z, f'({x}, {y}, {z})', color='red', fontsize=12)

    # Mostrar el gráfico
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Vectores originales
    vector1 = [3, 4, 5]
    vector2 = [2, 8, 5]

    # Sumar los vectores
    vector_suma = [vector1[0] + vector2[0], vector1[1] + vector2[1], vector1[2] + vector2[2]]

    # Representar el vector original 1 en color azul
    representar_vector(vector1, 'blue', 'Vector 1')

    # Representar el vector original 2 en color verde
    representar_vector(vector2, 'green', 'Vector 2')

    # Representar el vector suma en color rojo
    representar_vector(vector_suma, 'red', 'Vector Suma')
```
