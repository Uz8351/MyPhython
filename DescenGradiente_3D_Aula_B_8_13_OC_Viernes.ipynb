```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define la función f(x, y)
def func(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

# Inicializa el punto de partida
initial_point = np.array([18, 12])

# Realiza el descenso de gradiente usando el método de BFGS
result = minimize(func, initial_point, method='BFGS', options={'disp': True})

# Muestra el resultado
print("Resultado del descenso de gradiente:")
print(result)

# Extrae los valores de x e y después de la optimización
x_opt, y_opt = result.x

# Crea una malla de puntos para graficar la función
x = np.linspace(-5, 25, 400)
y = np.linspace(-5, 25, 400)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

# Grafica la función y el camino del descenso de gradiente
plt.contour(X, Y, Z, levels=20)
plt.plot(result.x[0], result.x[1], 'ro')  # Marca el punto óptimo con un círculo rojo
plt.title("Descenso de Gradiente")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define la función f(x, y)
def func(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

# Inicializa el punto de partida
initial_point = np.array([13, 12])

# Realiza el descenso de gradiente usando el método de BFGS
path = [initial_point]  # Lista para almacenar los puntos del camino
result = minimize(func, initial_point, method='BFGS', options={'disp': True}, callback=lambda xk: path.append(xk))

# Extrae los valores de x e y después de la optimización
x_opt, y_opt = result.x

# Crea una malla de puntos para graficar la función
x = np.linspace(-20, 20, 400)
y = np.linspace(-20, 20, 400)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

# Configura el tamaño de la figura para hacerlas más grandes
fig = plt.figure(figsize=(13, 12))

# Grafica la función en 3D
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Superficie 3D de la función')

# Extrae las coordenadas del camino
path = np.array(path).T

# Marca el camino del descenso de gradiente en la superficie 3D
ax.plot(path[0], path[1], func(path), 'ro-')

# Grafica las líneas de nivel en planta
plt.figure(figsize=(14, 12))  # Tamaño de la segunda figura
contour = plt.contour(X, Y, Z, levels=20, colors='k')
plt.clabel(contour, inline=1, fontsize=10)
plt.plot(path[0], path[1], 'ro-')  # Marca el camino con puntos rojos
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Líneas de nivel en planta')

plt.show()
```
