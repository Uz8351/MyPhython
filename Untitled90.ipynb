```python
from keras.engine import sequential
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
entradas = np.array([12, 14, 10, 13, 15, 16, 18, 15])
salidas = np.array([20, 34, 93, 85, 78, 82, 18, 88])

"""
En el contexto de las redes neuronales, Sequential()
es una clase en la biblioteca Keras que se utiliza
para construir modelos de redes neuronales secuenciales.

Un modelo secuencial es aquel en el que las capas
de la red neuronal se apilan una encima de la otra,
formando una secuencia lineal de capas.
Cada capa está conectada directamente a la siguiente
capa en la secuencia, y la información fluye
en una dirección, de la entrada a la salida.

La clase Sequential() de Keras nos permite
construir fácilmente modelos secuenciales.
Podemos agregar capas a este modelo utilizando
el método add(), que toma como argumento una
instancia de la capa que deseamos agregar.
Por ejemplo, model.add(Dense(4, input_dim=8, activation='relu'))
agrega una capa densa con 4 neuronas ocultas,
una dimensión de entrada de 8 y una función de activación ReLU.

El uso de Sequential() hace que la construcción
de modelos secuenciales sea más sencilla y legible.
Sin embargo, es importante tener en cuenta
que esta clase es adecuada para modelos secuenciales simples.
Si necesitas construir modelos más complejos
con múltiples ramificaciones o conexiones,
Keras ofrece otras opciones, como la API funcional.

En resumen, Sequential() es una clase
en Keras que permite construir
fácilmente modelos de redes neuronales
secuenciales, donde las capas se apilan
una encima de la otra en una secuencia lineal.
DENSE.
La capa Dense se llama así porque cada neurona
en esta capa está conectada a todas las neuronas
de la capa anterior y a todas las neuronas de
la capa siguiente. En otras palabras, cada neurona
recibe entradas de todas las neuronas de la capa
anterior y envía sus salidas a todas las neuronas
de la capa siguiente.
Cálculo ponderado: Cada conexión entre las neuronas
se representa mediante un peso. Durante el
cálculo ponderado, la capa Dense
multiplica las entradas de cada neurona por su peso correspondiente.
Cuando se crea una capa Dense en una red neuronal,
se debe especificar el número de neuronas o unidades
en esa capa. Por ejemplo, Dense(4) crea una capa
Dense con 4 neuronas. Además, se pueden especificar
otros argumentos opcionales, como la función de
activación a utilizar, la inicialización de pesos y los sesgos.
SIGMOID
f(x) = 1 / (1 + e^(-x))
Función en S, rango 0 a 1.
La función sigmoide tiene la propiedad de transformar
cualquier valor de entrada en un valor en el rango de 0 a 1.
Cuando x es negativo, la función sigmoide se acerca a 0,
y cuando x es positivo, se acerca a 1.
El punto medio, x = 0, produce una salida de 0.5.
"""

model.add(Dense(4, input_dim=1, activation='relu'))


#Función relu f(x) = max(0, x)
#la función ReLU toma el valor de entrada x
#y devuelve x si x es mayor o igual a cero, y devuelve cero en caso contrario.

model.add(Dense(2, activation='sigmoid'))
  # Capa de salida con 2 neuronas
# Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

"""
La línea de código model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
se utiliza para compilar un modelo de red neuronal
en Keras. Veamos cada uno de los argumentos utilizados en esta línea:

loss='binary_crossentropy': Este argumento especifica la función
de pérdida que se utilizará durante el entrenamiento del modelo.
En este caso, se utiliza la función de pérdida "binary_crossentropy",
que es comúnmente utilizada en problemas de clasificación binaria.
Esta función de pérdida es adecuada cuando se desea calcular
la discrepancia entre las salidas reales y las salidas
predichas para una clasificación binaria.

optimizer='adam': Este argumento especifica el optimizador
que se utilizará para ajustar los pesos y los sesgos del
modelo durante el entrenamiento. En este caso,
se utiliza el optimizador "adam", que es un optimizador popular
y eficiente en términos de rendimiento.
El optimizador Adam se adapta de forma adaptativa las
tasas de aprendizaje durante el entrenamiento y
puede ayudar a acelerar la convergencia del modelo.

metrics=['accuracy']: Este argumento especifica
las métricas que se calcularán y mostrarán durante
el entrenamiento y la evaluación del modelo.
En este caso, se utiliza la métrica "accuracy" (exactitud),
que es una medida comúnmente utilizada para evaluar el
rendimiento de un modelo de clasificación.
La exactitud proporciona la proporción de ejemplos
 clasificados correctamente sobre el total de ejemplos.

Después de compilar el modelo con estos argumentos,
el modelo está listo para ser entrenado utilizando
los datos de entrenamiento proporcionados en el
siguiente paso, utilizando el método fit().

Es importante tener en cuenta que la elección
de la función de pérdida, el optimizador y las
métricas puede variar según el problema y
los requisitos específicos. En función de la
naturaleza del problema y los datos, es posible
que desees explorar diferentes opciones
para obtener los mejores resultados.

"""
# Entrenar el modelo
model.fit(entradas, salidas, epochs=100, batch_size=12)

# Realizar predicciones
predicciones = model.predict(entradas)

# Imprimir las predicciones por consola
for prediccion in predicciones:
    print(prediccion)


```
