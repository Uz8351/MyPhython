```python
import numpy as np
import matplotlib.pyplot as plt

# Define la función f(x, y)
def f(x, y):
    return -4 * x / (x**2 + y**2 + 1)

# Define la función de costo (MSE)
def mse(x, y, target):
    return (f(x, y) - target) ** 2

# Inicialización de parámetros
x = -1
y = 0
learning_rate = 0.04
num_iterations = 38

# Listas para guardar los valores de x, y y costo en cada iteración
x_values = [x]
y_values = [y]
cost_values = []

# Descenso de gradiente
for i in range(num_iterations):
    gradient_x = -8 * x / (x**2 + y**2 + 1)**2
    gradient_y = 8 * y * x / (x**2 + y**2 + 1)**2
    x -= learning_rate * gradient_x
    y -= learning_rate * gradient_y
    cost = mse(x, y, 1.5)

    x_values.append(x)
    y_values.append(y)
    cost_values.append(cost)

# Crear la malla para el relieve 3D
x_mesh = np.linspace(-5, 5, 100)
y_mesh = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_mesh, y_mesh)
Z = f(X, Y)
print("********************************")
print("Valores costos MSE, por iteración:")
for index, miscostos in enumerate(cost_values):
    print(f'Íteración {index}: {miscostos}')
print("********************")
print(f'Punto óptimo después del descenso de gradiente: x={x}, y={y}')
print(f'Valor óptimo de f(x, y): {f(x, y)}')
print("*************************************************************")

# Crear las gráficas
fig = plt.figure(figsize=(20, 8))

# Gráfica del relieve 3D
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.set_zlabel('f(x, y)')

# Gráfica de líneas de nivel
ax2 = fig.add_subplot(132)
contours = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.clabel(contours, inline=1, fontsize=10)
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')

# Gráfica de costos del descenso de gradiente (MSE)
ax3 = fig.add_subplot(133)
ax3.plot(cost_values,color='black',marker='o',markerfacecolor='red')
ax3.set_xlabel('Iteración')
ax3.set_ylabel('Costo (MSE)')

plt.show()

```
