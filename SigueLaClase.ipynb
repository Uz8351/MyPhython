```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funcin y sus derivadas
def f(x):
    return x**2 + x

def df(x):
    return 2*x + 1

# Parmetros de la optimizacin
learning_rate = 0.05
iterations = 40
start_x = 18

# Inicializar listas para almacenar los valores de x y f(x) en cada iteracin
x_values = [start_x]
f_values = [f(start_x)]

# Descenso de gradiente
for _ in range(iterations):
    gradient = df(x_values[-1])
    new_x = x_values[-1] - learning_rate * gradient
    x_values.append(new_x)
    f_values.append(f(new_x))
for x in x_values:
    print(x)


# Crear puntos rojos en la grfica para mostrar el descenso
red_points_x = x_values[:-1]  # Excluimos el ltimo punto para marcar el punto mnimo
red_points_y = f_values[:-1]

# Pintar la grfica
x_range = np.linspace(-20, 20, 400)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = x^2 + x')
plt.scatter(red_points_x, red_points_y, color='red', label='Descenso de gradiente')
plt.scatter(start_x, f(start_x), color='green', label='Punto inicial')
plt.scatter(x_values[-1], f_values[-1], color='blue', label='Punto final')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente en f(x) = x^2 + x')
plt.grid(True)
plt.legend()
plt.show()
```
