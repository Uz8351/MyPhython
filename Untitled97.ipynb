```python
import numpy as MiMatriz
import matplotlib.pyplot as Graficacion

def crear_matriz_3d():
    matriz = MiMatriz.zeros((3, 3, 3))
    return matriz

def ingresar_datos(matriz):
    for i in range(27):
        while True:
            try:
                valor = float(input(f"Ingresa el valor para el elemento {i+1}: "))
                break
            except ValueError:
                print("Error: Debes ingresar un valor numérico.")

        x = i // 9
        y = (i // 3) % 3
        z = i % 3
        matriz[x, y, z] = valor

def graficar_matriz(matriz):
    fig = Graficacion.figure()
    ax = fig.add_subplot(111, projection='3d')

    x_data, y_data, z_data = MiMatriz.indices((3, 3, 3))
    valores = matriz.flatten()

    ax.scatter(x_data.flatten(), y_data.flatten(), z_data.flatten(), c=valores)

    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')
    ax.set_zlabel('Eje Z')

    Graficacion.show()

# Crear la matriz 3D
matriz = crear_matriz_3d()

# Ingresar los datos por consola
ingresar_datos(matriz)

# Graficar la matriz
graficar_matriz(matriz)
```
