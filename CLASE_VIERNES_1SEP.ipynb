```python
import numpy as np
import matplotlib.pyplot as plt

# Función de costo
def cost_function(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    error = (predictions - y)
    cost = (1/(2*m)) * np.dot(error.T, error)
    return cost

# Gradiente
def gradient(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    error = (predictions - y)
    grad = (1/m) * X.T.dot(error)
    return grad

# Datos
X = np.array([[1, 2], [3, 4], [5, 6],[3, 2], [7, 4], [5, 8]])
y = np.array([4, 7, 10.5, 9, 4,9])

# Theta inicial
theta = np.array([0, 0])  # Theta es un array unidimensional

# Hiperparámetros
learning_rate = 0.01
iterations = 1000

# Historial de costo
cost_history = np.zeros(iterations)

# Historial de theta para la línea de descenso
theta_history = []

# Descendiendo gradiente
for i in range(iterations):
    cost = cost_function(X, y, theta)
    cost_history[i] = cost
    grad = gradient(X, y, theta)
    theta = theta - (learning_rate * grad)
    theta_history.append(theta.copy())  # Guardar el valor de theta en esta iteración

# Gráfica
plt.scatter(X[:, 0], y, label='Datos de entrenamiento')
plt.plot(X[:, 0], X.dot(theta), color='red', label='Línea de descenso de gradiente')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Descenso de gradiente batch')
plt.legend()
plt.show()

plt.plot(range(iterations), cost_history)
plt.xlabel('Iteraciones')
plt.ylabel('Costo')
plt.title('Convergencia del descenso de gradiente batch')
plt.show()

print(f"Theta final: {theta}")
```
