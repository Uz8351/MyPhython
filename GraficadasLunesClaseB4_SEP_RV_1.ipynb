```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de muestra sintéticos
np.random.seed(0)
X = 2 * np.random.rand(10, 1)
y = 4 + 3 * X + np.random.randn(10, 1)
print("X:\n:",X)
print("*************")
print("y:\n:",y)

# Inicializar parámetros m y b
m = 0
b = 0

# Hiperparámetros del descenso de gradiente
learning_rate = 0.01
iterations = 8  # Aumenté el número de iteraciones para una mejor convergencia

# Listas para guardar los valores de pérdida y parámetros en cada iteración
losses = []
ms = []
bs = []

# Descenso de gradiente tipo Batch
for i in range(iterations):
    # Calcular las predicciones del modelo
    y_pred = m * X + b

    # Calcular el gradiente de la función de costo (MSE)
    gradient_m = np.mean((y_pred - y) * X)
    gradient_b = np.mean(y_pred - y)
    print("gradient_m:\n",gradient_m)
    print("gradient_b:\n",gradient_b)


    # Actualizar los parámetros m y b
    m = m - learning_rate * gradient_m
    b = b - learning_rate * gradient_b

    # Calcular y almacenar la pérdida en esta iteración
    loss = np.mean((y_pred - y) ** 2)
    losses.append(loss)
    ms.append(m)
    bs.append(b)


print("************************")
print("y_pred:\n",y_pred)
#---------------------------------------------------------
# Asumiendo que las tres listas tienen la misma longitud
for i in range(len(losses)):
    print(f"iterations {i + 1}: losses={losses[i]}, ms={ms[i]}, bs={bs[i]}")

# Hacer predicciones en nuevos datos
new_x = np.array([[1.5]])  # Nuevo valor de x
new_y = m * new_x + b

# Graficar la pérdida en función de las iteraciones y la regresión lineal en la misma figura
plt.figure(figsize=(15, 7))

# Subtrama 1: Pérdida en función de las iteraciones
plt.subplot(1, 2, 1)
plt.plot(range(iterations), losses, label='Pérdida', color='blue')
plt.scatter(range(iterations), losses, c='red', marker='o', label='Valores de Pérdida')
plt.xlabel('Iteraciones')
plt.ylabel('Pérdida')
plt.title('Pérdida en función de las Iteraciones')
plt.legend()

# Subtrama 2: Regresión lineal con descenso de gradiente
plt.subplot(1, 2, 2)
plt.scatter(X, y, label='Datos de muestra')
plt.plot(X, m * X + b, color='red', label='Línea de regresión')
plt.scatter(new_x, new_y, color='green', marker='o', label='Predicción')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión lineal con descenso de gradiente')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Imprimir los parámetros finales del modelo
print(f"Parámetro m: {m}")
print(f"Parámetro b: {b}")
# Definir los parámetros de la función f(x) = mx + b


# Crear un rango de valores para x
x = np.linspace(-15, 15, 100)

# Calcular los valores de y usando la ecuación f(x) = mx + b
y = m * x + b

# Crear el gráfico
plt.figure(figsize=(15, 7))
plt.plot(x, y, label=f'f(x) = {m:.2f}x + {b:.2f}', color='black',linewidth=2.5)

# Agregar líneas punteadas para representar las iteraciones
for i, (mi, bi) in enumerate(zip(ms, bs)):
    yi = mi * x + bi
    plt.plot(x, yi, linestyle='--', label=f'Iteración {i + 1}: m={mi:.2f}, b={bi:.2f}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = mx + b con Iteraciones (Líneas Punteadas)')
plt.grid(True)
plt.legend()
plt.show()




```
