```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Definir la función
def f(x):
    return x**2/5

# Parámetros de regresión
x_inicial = 23
num_iteraciones = 120
learning_rate = 0.1  # Tasa de aprendizaje

# Inicializar valores
x_actual = x_inicial
historial_x = []
historial_y = []

# Realizar iteraciones
for _ in range(num_iteraciones):
    # Calcular la derivada utilizando diferencias finitas
    derivada = (f(x_actual) - f(x_actual - learning_rate)) / learning_rate

    # Actualizar el valor de x utilizando el gradiente descendiente
    x_actual = x_actual - learning_rate * derivada

    # Guardar los valores de la iteración en el historial
    historial_x.append(x_actual)
    historial_y.append(f(x_actual))

# Imprimir el resultado
valor_minimo = f(x_actual)
print("El valor mínimo estimado es:", valor_minimo)

# Crear una lista de puntos para la gráfica
x_valores = np.linspace(1, 30, 400)
y_valores = f(x_valores)

# Graficar la función, los puntos de iteración y la línea de regresión
fig, ax = plt.subplots()
ax.plot(x_valores, y_valores, label='f(x)')
ax.scatter(historial_x, historial_y, color='red', label='Puntos de iteración')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Mínimo valor por derivada y regresión')

# Configurar grid reducido en las cuadrículas menores
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.8))
ax.yaxis.set_major_locator(ticker.MultipleLocator(18))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(5))

ax.grid(which='both', linestyle='--', linewidth=0.5)

ax.legend()
plt.show()
```
