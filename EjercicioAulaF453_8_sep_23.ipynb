```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)

# Funcin para calcular el error cuadrtico medio (MSE)
def calcular_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# Inicializar parmetros (pendiente y ordenada al origen)
theta0 = np.random.randn()
theta1 = np.random.randn()

# Parmetros del descenso de gradiente
learning_rate = 0.1
num_iteraciones = 100
historial_costo = []

# Realizar el descenso de gradiente
for i in range(num_iteraciones):
    # Calcular las predicciones
    y_pred = theta0 + theta1 * X

    # Calcular el gradiente del MSE
    gradiente_theta0 = -2 * (y - y_pred).mean()
    gradiente_theta1 = -2 * (X * (y - y_pred)).mean()

    # Actualizar los parmetros
    theta0 = theta0 - learning_rate * gradiente_theta0
    theta1 = theta1 - learning_rate * gradiente_theta1

    # Calcular el costo (MSE) y agregarlo al historial
    costo = calcular_mse(y, y_pred)
    historial_costo.append(costo)

# Graficar la evolucin del costo (MSE)
plt.plot(historial_costo)
plt.xlabel('Iteraciones')
plt.ylabel('MSE')
plt.title('Evolucin del Error Cuadrtico Medio (MSE) durante el Descenso de Gradiente')
plt.show()

# Mostrar la lnea ajustada
plt.scatter(X, y, label='Datos')
plt.plot(X, theta0 + theta1 * X, color='red', label='Ajuste Lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste Lineal usando Descenso de Gradiente (MSE)')
plt.show()

print(f'Parmetros finales: theta0 = {theta0}, theta1 = {theta1}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)

# Funcin para calcular el error cuadrtico medio (MSE)
def calcular_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# Inicializar parmetros (pendiente y ordenada al origen)
theta0 = np.random.randn()
theta1 = np.random.randn()

# Parmetros del descenso de gradiente
learning_rate = 0.1
num_iteraciones = 100
historial_costo = []
historial_theta0 = []
historial_theta1 = []

# Realizar el descenso de gradiente
for i in range(num_iteraciones):
    # Calcular las predicciones
    y_pred = theta0 + theta1 * X

    # Calcular el gradiente del MSE
    gradiente_theta0 = -2 * (y - y_pred).mean()
    gradiente_theta1 = -2 * (X * (y - y_pred)).mean()

    # Actualizar los parmetros
    theta0 = theta0 - learning_rate * gradiente_theta0
    theta1 = theta1 - learning_rate * gradiente_theta1

    # Calcular el costo (MSE) y agregarlo al historial
    costo = calcular_mse(y, y_pred)
    historial_costo.append(costo)

    # Agregar los parmetros al historial
    historial_theta0.append(theta0)
    historial_theta1.append(theta1)

# Graficar la evolucin del costo (MSE) con puntos rojos
plt.plot(historial_costo, label='MSE')
plt.scatter(range(num_iteraciones), historial_costo, c='red', marker='o', label='Saltos del Descenso')
plt.xlabel('Iteraciones')
plt.ylabel('MSE')
plt.title('Evolucin del Error Cuadrtico Medio (MSE) durante el Descenso de Gradiente')
plt.legend()
plt.show()

# Mostrar la lnea ajustada
plt.scatter(X, y, label='Datos')
plt.plot(X, theta0 + theta1 * X, color='red', label='Ajuste Lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste Lineal usando Descenso de Gradiente (MSE)')
plt.show()

print(f'Parmetros finales: theta0 = {theta0}, theta1 = {theta1}')
```
