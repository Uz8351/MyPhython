```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios para entrada y salida
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Funcin para calcular el error cuadrtico medio (MSE)
def calcular_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# Inicializar parmetros del modelo (pendiente y sesgo)
theta = np.random.randn(2)

# Hiperparmetros del descenso de gradiente
lr = 0.1  # Tasa de aprendizaje
n_iter = 100  # Nmero de iteraciones

# Almacenar el error cuadrtico medio en cada iteracin
mse_history = []

# Realizar el descenso de gradiente
for iteracion in range(n_iter):
    # Calcular predicciones del modelo
    y_pred = X.dot(theta[1]) + theta[0]

    # Calcular el gradiente del error con respecto a los parmetros
    gradiente = -2 * X.T.dot(y - y_pred) / len(y)

    # Actualizar los parmetros usando la regla del descenso de gradiente
    theta[0] = theta[0] - lr * gradiente[0]
    theta[1] = theta[1] - lr * gradiente[0]  # Usa el mismo gradiente para ambos parmetros

    # Calcular el error cuadrtico medio y almacenarlo
    mse = calcular_mse(y, y_pred)
    mse_history.append(mse)

# Crear un rango de valores para la lnea de descenso
x_line = np.linspace(0, 2, 100)
y_line = theta[0] + theta[1] * x_line

# Graficar los puntos de datos en rojo
plt.scatter(X, y, color='red', label='Datos')

# Graficar la lnea de descenso en azul
plt.plot(x_line, y_line, color='blue', label='Lnea de Descenso')

# Configurar el grfico
plt.xlabel('X')
plt.ylabel('y')
plt.title('Descenso de Gradiente Batch - Ajuste Lineal')
plt.legend()

# Mostrar la grfica
plt.show()

# Graficar la evolucin del error cuadrtico medio
plt.plot(range(n_iter), mse_history)
plt.xlabel('Iteracin')
plt.ylabel('MSE')
plt.title('Evolucin del Error Cuadrtico Medio durante el Descenso de Gradiente')
plt.show()
```
