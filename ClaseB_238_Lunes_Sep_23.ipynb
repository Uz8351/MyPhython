```python
"""
ClaseB-238_Lunes_Sep_23.ipynb
-----------------------------------
Hilario Iglesias Martnez
*************************************
Este es un programa en Python
se realiza una regresin lineal
utilizando el mtodo de gradiente descendente
(Batch Gradient Descent)
El programa realiza una regresin lineal utilizando
el mtodo de gradiente descendente
y visualiza los resultados a travs de
grficas. Es una implementacin simple
pero efectiva de un modelo de regresin lineal.
"""

#Importacin de biblioteca:NumPy realiza operaciones matriciales.
#Importacin de Matplotlib para la creacin de grficos.
import numpy as np
import matplotlib.pyplot as plt

# Datos aleatorios de ejemplo.
"""
Se generan datos de ejemplo Aleatorios_X, y
Aleatorios_y. Utilizando la funcin np.random.rand de NumPy.
Estos datos se utilizan para entrenar el modelo de regresin.
"""
np.random.seed(0)
Aleatorios_X = 2 * np.random.rand(150, 1)
Aleatorios_y = 4 + 3 * Aleatorios_X + np.random.randn(150, 1)

# Inicializacin de parmetros (pendiente y ordenada al origen)
Pendiente = 0
Intercepto = 0

#Parmetros de Tasa de aprendizaje, y Iteraciones o epoch.
Tasa_Aprendizaje=0.01
Iteraciones = 100

# Listas para guardar los valores de prdida y parmetros en cada iteracin.
"""
Perdidas, Pendientes, e Interceptos
son listas vacas que se utilizarn
para almacenar los valores de prdida,
y los parmetros en cada iteracin.
"""
Perdidas = []
Pendientes = []
Interceptos = []# epoch

# Bucle de iteracin
"""
Se ejecuta un bucle for que itera a travs del
nmero de iteraciones especificado.
En cada iteracin, se calculan las predicciones (Predicciones)
utilizando la ecuacin de regresin lineal.
Se calcula el error entre las predicciones y los valores reales.
Se calculan los gradientes (Gradiente_Pendiente y Gradiente_Intercepto)
utilizando el gradiente descendente para actualizar los parmetros
(Pendiente e Intercepto).
Se calcula la funcin de prdida, que es el error cuadrtico medio.
Los valores de prdida, pendiente e intercepto se almacenan en
las listas correspondientes para su posterior visualizacin.

"""

for epoch in range(Iteraciones):
    # Clculo de las predicciones
    Predicciones = Pendiente * Aleatorios_X + Intercepto

    # Clculo del error
    Error = Aleatorios_y - Predicciones

    # Clculo de los gradientes
    Gradiente_Pendiente = (-2 / len(Aleatorios_X)) * np.sum(Aleatorios_X * Error)
    Gradiente_Intercepto = (-2 / len(Aleatorios_X)) * np.sum(Error)

    # Actualizacin de parmetros
    Pendiente -= Tasa_Aprendizaje * Gradiente_Pendiente
    Intercepto -= Tasa_Aprendizaje * Gradiente_Intercepto

    # Clculo de la funcin de prdida (Error cuadrtico medio)
    Perdida = np.mean(Error ** 2)

    # Almacenar valores para graficar
    Perdidas.append(Perdida)
    Pendientes.append(Pendiente)
    Interceptos.append(Intercepto)

# Parmetros finales
Final_Pendiente = Pendientes[-1] if Pendientes else 0
Final_Intercepto = Interceptos[-1] if Interceptos else 0
#******************************************************
"""
Se realizan tres predicciones (nuevo_x, nuevo1_x, y nuevo2_x)
utilizando la pendiente e intercepto finales obtenidos despus
de todas las iteraciones.
"""
m = Pendiente
b = Intercepto
nuevo_x = np.array([[0.5]])  # Nuevo valor de x
nuevo_y = m * nuevo_x + b
nuevo1_x = np.array([[0.8]])  # Nuevo valor de x
nuevo1_y = m * nuevo1_x + b
nuevo2_x = np.array([[0.6]])  # Nuevo valor de x
nuevo2_y = m * nuevo2_x + b
#*****************************************************
# Grfica de la prdida en funcin de las Iteraciones.
"""
Se crea una figura con dos subgrficos:
El primer subgrfico muestra la prdida
en funcin de las iteraciones.
El segundo subgrfico muestra los datos reales,
la lnea de regresin ajustada y las predicciones realizadas.
"""
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, Epocas + 1), Perdidas, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Prdida')
plt.title('Prdida en funcin de las Iteraciones')

# Grfica de la regresin lineal ajustada
plt.subplot(1, 2, 2)
plt.scatter(Aleatorios_X, Aleatorios_y, label='Datos reales')
plt.plot(Aleatorios_X, Final_Pendiente * Aleatorios_X + Final_Intercepto, color='red',
label='Regresin lineal,m: 3.4501,b: 3.2805')

plt.scatter(nuevo_x, nuevo_y, color='black', marker='o', s=100, label='Prueba de prediccin_0')
plt.scatter(nuevo1_x, nuevo1_y, color='black', marker='o', s=100, label='Prueba de prediccin_1')
plt.scatter(nuevo2_x, nuevo2_y, color='black', marker='o', s=100, label='Prueba de prediccin_2')



plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.legend()
plt.title('Regresin Lineal')

plt.tight_layout()
plt.show()

#Impresin de parmetros finales:
print("*************************************************")
print(f"Pendiente final m: {Final_Pendiente:.4f}")
print(f"Ordenada al origen final b: {Final_Intercepto:.4f}")
print("*************************************************")
```
