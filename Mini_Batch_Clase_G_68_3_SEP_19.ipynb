```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo: f(x) = x^2
def objective_function(x):
    return x**2

# Derivada de la función objetivo: f'(x) = 2x
def gradient(x):
    return 2 * x

# Descenso de gradiente en mini lotes
def mini_batch_gradient_descent(learning_rate, batch_size, epochs):
    x = np.random.rand(1) * 10  # Inicialización aleatoria de x
    history = []  # Almacenar historial de valores de x
    for epoch in range(epochs):
        for _ in range(0, len(x), batch_size):
            x_batch = x[_:_+batch_size]  # Mini lote
            gradient_batch = gradient(x_batch)  # Gradiente del mini lote
            x -= learning_rate * gradient_batch.mean()  # Actualización de x
            history.append(x[0])
    return history

# Hiperparámetros
learning_rate = 0.1
batch_size = 5
epochs = 50

# Ejecutar descenso de gradiente en mini lotes
history = mini_batch_gradient_descent(learning_rate, batch_size, epochs)

# Gráfico de la convergencia
plt.figure(figsize=(10, 6))
plt.plot(range(len(history)), [objective_function(x) for x in history], label='Función Objetivo')
plt.xlabel('Iteración')
plt.ylabel('Valor de la Función Objetivo')
plt.legend()
plt.title('Descenso de Gradiente en Mini Lotes')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generación de datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # Modelo verdadero + ruido

# Parámetros iniciales del modelo
theta0 = np.random.randn()
theta1 = np.random.randn()

# Hiperparámetros
learning_rate = 0.01
batch_size = 10
epochs = 100

# Descenso de gradiente en mini lotes
history_theta0 = []
history_theta1 = []

for epoch in range(epochs):
    for _ in range(0, len(X), batch_size):
        X_batch = X[_:_+batch_size]
        y_batch = y[_:_+batch_size]
        n = len(X_batch)

        # Cálculo de predicciones y error
        predictions = theta0 + theta1 * X_batch
        error = predictions - y_batch

        # Actualización de parámetros
        gradient_theta0 = (1/n) * np.sum(error)
        gradient_theta1 = (1/n) * np.sum(error * X_batch)
        theta0 -= learning_rate * gradient_theta0
        theta1 -= learning_rate * gradient_theta1

        history_theta0.append(theta0)
        history_theta1.append(theta1)

# Gráfico de ajuste lineal final
plt.figure(figsize=(12, 6))
plt.scatter(X, y, label='Datos de entrenamiento')
plt.plot(X, theta0 + theta1 * X, color='red', label='Ajuste Lineal Final')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste Lineal con Descenso de Gradiente en Mini Lotes')
plt.show()

# Gráfico de cómo evolucionan los parámetros theta0 y theta1
plt.figure(figsize=(12, 6))
plt.plot(range(len(history_theta0)), history_theta0, label='Theta0')
plt.plot(range(len(history_theta1)), history_theta1, label='Theta1')
plt.xlabel('Iteración')
plt.ylabel('Valor de Parámetro')
plt.legend()
plt.title('Evolución de los Parámetros en el Descenso de Gradiente en Mini Lotes')
plt.grid(True)
plt.show()
```
