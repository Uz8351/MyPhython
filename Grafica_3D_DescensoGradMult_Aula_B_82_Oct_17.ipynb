```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Función f(x, y)
def f(x, y):
    return 4 * x / (x**2 + y**2 + 1)

# Crear una malla de puntos en el espacio x-y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Crear la figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear la superficie 3D
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Punto en rojo
x_point = -0.92
y_point = 0.04
z_point = 1.9913
ax.scatter([x_point], [y_point], [z_point], c='red', s=50)  # Añadir el punto en rojo

# Añadir una barra de colores
fig.colorbar(surf)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')

# Título
plt.title('Gráfico 3D de f(x, y) con el punto en rojo')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función f(x, y)
def f(x, y):
    return 4 * x / (x**2 + y**2 + 1)

# Gradiente de la función f(x, y)
def gradient(x, y):
    df_dx = (4 * (1 - x**2 - y**2)) / (x**2 + y**2 + 1)**2
    df_dy = (-8 * x * y) / (x**2 + y**2 + 1)**2
    return df_dx, df_dy

# Descenso de gradiente
def gradient_descent(x, y, learning_rate, num_iterations):
    x_history = [x]
    y_history = [y]
    cost_history = []
    for _ in range(num_iterations):
        df_dx, df_dy = gradient(x, y)
        x -= learning_rate * df_dx
        y -= learning_rate * df_dy
        x_history.append(x)
        y_history.append(y)
        cost = (f(x, y) - 1.9913) ** 2  # MSE cost
        cost_history.append(cost)
    return x_history, y_history, cost_history

# Punto inicial
x_initial = 0.92
y_initial = 0.16

ax.scatter([x_initial], [y_initial], [z_initial], c='red', s=50, label='Punto inicial (0.92, 0.16, 1.9658)')

# Parámetros del descenso de gradiente
learning_rate = 0.1
num_iterations = 10

x_history, y_history, cost_history = gradient_descent(x_initial, y_initial, learning_rate, num_iterations)
print("*******************************")
for index, miscostos in enumerate(cost_history):
  print(f'Interación {index}:{miscostos}')
print("*******************************")
# Gráfica del relieve 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()

# Gráfica de líneas de nivel
plt.figure()
CS = plt.contour(X, Y, Z, levels=np.arange(0, 2, 0.1), colors='black')
plt.clabel(CS, inline=1, fontsize=10)

# Gráfica de la función de costo (MSE)
plt.figure()
plt.plot(cost_history,color='black',marker='o',markerfacecolor='red')
plt.xlabel('Iteraciones')
plt.ylabel('Costo (MSE)')

plt.show()
```
