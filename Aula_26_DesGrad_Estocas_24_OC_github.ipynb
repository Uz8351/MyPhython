```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo f(x) = log2(x)
def f(x):
    return np.log2(x)

# Derivada de la función
def df(x):
    return 1 / (x * np.log(2))

# Descenso estocástico
def stochastic_descent(learning_rate, num_iterations, start_x):
    x_values = [start_x]
    costs = []

    for _ in range(num_iterations):
        gradient = df(x_values[-1])
        x_new = x_values[-1] - learning_rate * gradient
        cost = (f(x_values[-1]) - f(x_new)) ** 2
        x_values.append(x_new)
        costs.append(cost)

    return x_values, costs

learning_rate = 0.01
num_iterations = 10
start_x=18


# Definir el rango de valores de x
x = np.linspace(1, 30, 400)

x_values, costs = stochastic_descent(learning_rate, num_iterations, start_x)
print("*******************************")
print("Historial de Costo:")
for index, miscostos in enumerate(costs):
    print(f'Iteración {index}:{miscostos}')
print("*******************************")

# Crear una figura con ejes personalizados
fig, ax = plt.subplots()
ax.plot(x, f(x), label='f(x) = log2(x)')
ax.scatter(x_values, [f(x) for x in x_values], c='red', label='Descenso estocástico')

# Ajustar los límites de los ejes x e y
ax.set_xlim(1, 40)
ax.set_ylim(-8, 8)  # Ajusta estos valores según tus necesidades

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.set_title('Descenso Estocástico de f(x) = log2(x)')
plt.show()

# Graficar los costos (MSE) hasta el mínimo costo
plt.plot(range(1, num_iterations + 1), costs, marker='o', color='black')
plt.xlabel('Iteración')
plt.ylabel('Costo (MSE)')
plt.title('Evolución de Costos durante el Descenso Estocástico')
plt.show()
```
