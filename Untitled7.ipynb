```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
import numpy as np
import os
from PIL import Image
from io import BytesIO
import imghdr  # Importar la biblioteca para determinar el tipo de imagen

# Cargar el modelo InceptionV3 preentrenado en ImageNet
base_model = InceptionV3(weights='imagenet')

# Cambiar la ruta a la carpeta que contiene tus imágenes CIFAR-10
ruta_imagenes = '/content/drive/MyDrive/cifar-10-batches-py'

for imagen_nombre in os.listdir(ruta_imagenes):
    imagen_path = os.path.join(ruta_imagenes, imagen_nombre)

    # Verificar el tipo de imagen
    tipo_imagen = imghdr.what(imagen_path)

    if tipo_imagen is not None:
        # Cargar y preprocesar la imagen binaria
        with open(imagen_path, "rb") as f:
            img_data = BytesIO(f.read())
            img = Image.open(img_data)
            img = img.resize((299, 299))  # Ajusta el tamaño según sea necesario
            img_array = np.expand_dims(np.array(img), axis=0)
            img_array = preprocess_input(img_array)

        # Obtener predicciones
        predictions = base_model.predict(img_array)

        # Decodificar y mostrar las 3 principales predicciones
        decoded_predictions = decode_predictions(predictions, top=3)[0]
        print(f"Predicciones para {imagen_nombre}:")
        for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
            print(f"{i + 1}: {label} ({score:.2f})")

        print("\n")
    else:
        print(f"Tipo de imagen no válido para {imagen_nombre}")
```
