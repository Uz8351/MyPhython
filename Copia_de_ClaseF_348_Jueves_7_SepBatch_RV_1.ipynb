```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Inicializamos los parámetros del modelo
theta = np.random.randn(2, 1)

# Hiperparámetros
learning_rate = 0.01  # Valor más pequeño
n_iterations = 1000

# Almacenamos la historia de pérdida para visualización
loss_history = []

# Agregamos una columna de unos a la matriz X para el término de sesgo (bias)
X_b = np.c_[np.ones((100, 1)), X]

# Creamos una figura con un tamaño específico

plt.figure(figsize=(12, 6))

# Graficamos los puntos generados aleatoriamente
plt.scatter(X, y, label='Datos de entrenamiento')

for iteration in range(n_iterations):
    # Calculamos las predicciones del modelo
    predictions = X_b.dot(theta)

    # Calculamos el error
    error = predictions - y

    # Calculamos el gradiente de la función de costo (MSE)
    gradient = -2/len(X) * X_b.T.dot(error)

    # Actualizamos los parámetros theta usando el descenso de gradiente
    theta -= learning_rate * gradient

    # Calculamos el valor de la función de costo (MSE) y lo agregamos al historial
    mse = np.mean(error**2)
    loss_history.append(mse)
```
