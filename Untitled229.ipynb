```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return x**2 / 3

def df(x):
    return 2 * x / 3

# Parámetros de descenso de gradiente
x_init = 40
alpha = 0.1
num_iterations = 12

# Descenso de gradiente
x_values = [x_init]
for _ in range(num_iterations):
    x_new = x_values[-1] - alpha * df(x_values[-1])
    x_values.append(x_new)
    print(x_new)
# Crear una malla de puntos para la gráfica
x_range = np.linspace(-1, 15, 300)  # Generar valores equidistantes

# Graficar la función y el descenso de gradiente
plt.plot(x_range, f(x_range), label='f(x) = x^2 / 3')
plt.scatter(x_values, f(np.array(x_values)), color='black', label='Descenso de Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Ajustar la cuadrícula y las minor ticks en ambos ejes
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='red')
plt.minorticks_on()

plt.tight_layout()

# Imprimir el último valor más pequeño de la derivada (x_new)
x_new = x_values[-1] - alpha * df(x_values[-1])
print("Último valor más pequeño de la derivada:", x_new)

plt.show()
```
