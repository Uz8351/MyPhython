```python
from google.colab import drive
drive.mount('/content/drive')
#/content/drive/MyDrive/druida/flores
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Ruta al directorio que contiene subdirectorios de clases con imágenes de flores
data_dir = '/content/drive/MyDrive/druida/flores'

# Definir el tamaño de las imágenes y el tamaño del lote
img_size = (299, 299)
batch_size = 32

# Crear un generador de datos para cargar imágenes desde el directorio
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # Dividir el conjunto de datos en entrenamiento y validación
)

train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Cargar el modelo preentrenado InceptionV3 sin la capa superior (include_top=False)
base_model = InceptionV3(weights='imagenet', include_top=False)

# Añadir nuevas capas al modelo base
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

# Crear el modelo final
model = Model(inputs=base_model.input, outputs=predictions)

# Congelar las capas del modelo base
for layer in base_model.layers:
    layer.trainable = False

# Compilar el modelo
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
epochs = 10
model.fit(train_generator, epochs=epochs, validation_data=validation_generator)

# Guardar el modelo entrenado
model.save('modelo_transferencia_aprendizaje.h5')

# Realizar predicciones en una imagen de prueba
img_path = '/content/drive/MyDrive/druida/flores/flores1/image_00001.jpg'
img = tf.keras.preprocessing.image.load_img(img_path, target_size=img_size)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
img_array = preprocess_input(img_array)

predictions = model.predict(img_array)
decoded_predictions = decode_predictions(predictions, top=3)[0]

print("Predicciones:")
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}: {label} ({score:.2f})")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Ruta al directorio que contiene subdirectorios de clases con imágenes de flores
data_dir = '/content/drive/MyDrive/druida/flores'


# Definir el tamaño de las imágenes y el tamaño del lote
img_size = (299, 299)
batch_size = 32

# Crear un generador de datos para cargar imágenes desde el directorio
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # Dividir el conjunto de datos en entrenamiento y validación
)

train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Cargar el modelo preentrenado InceptionV3 sin la capa superior (include_top=False)
base_model = InceptionV3(weights='imagenet', include_top=False)

# Añadir nuevas capas al modelo base
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

# Crear el modelo final
model = Model(inputs=base_model.input, outputs=predictions)

# Congelar las capas del modelo base
for layer in base_model.layers:
    layer.trainable = False

# Compilar el modelo
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
epochs = 10
model.fit(train_generator, epochs=epochs, validation_data=validation_generator)

# Guardar el modelo entrenado
model.save('modelo_transferencia_aprendizaje.h5')

# Obtener las clases del generador de datos
class_labels = list(train_generator.class_indices.keys())

# Mapeo de clases botánicas
botanical_classes = {
    'flores1': 'Clase Botánica A',
    'flores2': 'Clase Botánica B',
    # Agrega más clases según sea necesario
}

# Realizar predicciones en una imagen de prueba
img_path = '/content/drive/MyDrive/druida/flores/flores1/image_00050.jpg'
img = tf.keras.preprocessing.image.load_img(img_path, target_size=img_size)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
img_array = preprocess_input(img_array)

predictions = model.predict(img_array)
predicted_class_index = tf.argmax(predictions[0]).numpy()
predicted_botanical_class = botanical_classes[class_labels[predicted_class_index]]

print("Predicción de Clase Botánica:")
print(predicted_botanical_class)
```
