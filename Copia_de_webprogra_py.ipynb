```python
"""
webprogra.py
Una función cualquiera de aplicación
para el ejenmplo.
f(x)=(π*x)/x**2+1
Se calculan para 18 Valores_Entrada, los Valores_Salida

Neuronas de entrada:1
Primera capa de neuronas intermedia oculta:20
Segunda capa de neuronas intermedia oculta:10
Capa de salida:1

"""

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

Valores_Entrada = np.array([0.12,0.3,0.5,1,2, 3, 4, 5, 6, 8, 9, 14, 18, 21, 22,30,40,50])
Valores_Salida = np.array([27.166,11.466,7.28,4.14,2.57, 2.046, 1.785, 1.523,1.628, 1.392, 1.348, 1.224, 1.174, 1.149, 1.427,1.104,1.0785,1.0628])

model = Sequential()
model.add(Dense(20, input_dim=1, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(Valores_Entrada, Valores_Salida, epochs=20000, verbose=0)
# Evaluación de el modelo
loss = model.evaluate(Valores_Entrada, Valores_Salida)
print('Loss:', loss)

plt.plot(Valores_Entrada, Valores_Salida, 'o')
plt.plot(Valores_Entrada, model.predict(Valores_Entrada), '-')
plt.xlabel('Valores entrada')
plt.ylabel('Valores salida')
plt.legend(['Datos de Entrenamiento', 'Linea Regresión lineal'], loc='best')
plt.show()

Valores_Salida_prediccion = model.predict(np.array([23]))
print('Prediccion de Salida:', Valores_Salida_prediccion[0])
```
