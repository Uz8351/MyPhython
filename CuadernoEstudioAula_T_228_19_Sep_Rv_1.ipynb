```python
import numpy as np
import matplotlib.pyplot as plt

# Hiperparámetros de SGD
learning_rate = 0.1
theta0 = 0.1
theta1 = 0.1
epochs = 20  # Agregamos el número de épocas

# Listas para almacenar el historial de costos y los parámetros
cost_history = []
theta0_history = []
theta1_history = []

# Datos de entrenamiento
np.random.seed(0)
X_train = 2 * np.random.rand(100, 1)
y_train = 4 + 3 * X_train + np.random.randn(100, 1)  # Incluye ruido

for epoch in range(epochs):
    mse_loss = np.mean((theta0 + theta1 * X_train - y_train) ** 2)

    # Cálculo de la predicción
    prediction = theta0 + theta1 * X_train

    n = len(X_train)
    grad_theta0 = (1/n) * np.sum(prediction - y_train)
    grad_theta1 = (1/n) * np.sum((prediction - y_train) * X_train)

    # Actualización de parámetros
    theta0 -= learning_rate * grad_theta0
    theta1 -= learning_rate * grad_theta1

    # Almacenar el historial de costos y parámetros
    cost_history.append(mse_loss)
    theta0_history.append(theta0)
    theta1_history.append(theta1)

    # Imprimir la pérdida en cada época
    print(f'Época {epoch+1}/{epochs}, Pérdida: {mse_loss}')

# Parámetros finales
print(f'Parámetro theta0: {theta0}')
print(f'Parámetro theta1: {theta1}')

# Gráfico de ajuste lineal
plt.figure(figsize=(12, 6))
plt.scatter(X_train, y_train, label='Datos de entrenamiento')
plt.plot(X_train, theta0 + theta1 * X_train, color='red', label='Ajuste Lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste Lineal con SGD y MSE')
plt.show()

# Gráfico del historial de costos
plt.figure(figsize=(12, 6))
plt.plot(range(epochs), cost_history, color='black', marker='o', markerfacecolor='red')
plt.xlabel('Época')
plt.ylabel('Pérdida (MSE)')
plt.title('Historial de Costos (MSE)')
plt.show()
```
