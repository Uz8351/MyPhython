```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def custom_function(x):
    return (x**2 / 3) * (x**2 / 5)

def custom_function_derivative(x):
    return 2*x**2/5

# Parámetros de la optimización
learning_rate = 0.01
iterations = 30000
x_initial = 16

# Listas para almacenar los puntos y valores durante las iteraciones
x_values = [x_initial]
y_values = [custom_function(x_initial)]

# Realizar iteraciones de descenso de gradiente
for _ in range(iterations):
    x_initial = x_initial - learning_rate * custom_function_derivative(x_initial)
    x_values.append(x_initial)
    y_values.append(custom_function(x_initial))
print("************************************************")

MinimoValorIndice=np.argmin(x_values)
MinimoEn_x=x_values[MinimoValorIndice]
MinimoEn_y=y_values[MinimoValorIndice]
print("Minimo valor de x:",MinimoEn_x)
print("Minimo valor en y:",custom_function(MinimoEn_x))

#print(MinimoEn_y)
print("************************************************")
# Crear la gráfica
x_range = np.linspace(-20, 20, 400)
y_range = custom_function(x_range)


plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Funcion:f(x)=(x**2 / 3) * (x**2 / 5)')
plt.scatter(x_values, y_values, color='red', label='Gradiente de descenso')
plt.scatter(x_values[0], y_values[0], color='green', label='Punto Inicial')
plt.scatter(x_values[-1], y_values[-1], color='blue', label='Punto final más proximo a valor 0')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(' Funcion, gradiente de descenso para:f(x)=(x**2 / 3) * (x**2 / 5)')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()

plt.show()
```
