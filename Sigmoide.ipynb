```python


import numpy as np
import matplotlib.pyplot as plt
e = 2.71828

# Definir la función sigmoide y su derivada
def sigmoid(x):
    return 1 / (1 + e**(-x))

def sigmoid_derivative(x):
    return e**(-x) / (1 + e**(-x))**2

# Crear un rango de valores a lo largo de la longitud de la función sigmoide
x_range = np.linspace(-10, 10, 400)  # Ajusta el rango según tus necesidades

# Calcular los valores de la función sigmoide y su derivada en el rango
sigmoid_values = sigmoid(x_range)
sigmoid_derivative_values = sigmoid_derivative(x_range)

# Parámetros iniciales para el proceso de derivada descendente
x_initial = 1.0
learning_rate = 0.1
iterations = 1000
x_history = [x_initial]
y_history = [sigmoid(x_initial)]

# Realizar el proceso de derivada descendente
for _ in range(iterations):
    x_initial = x_initial - learning_rate * sigmoid_derivative(x_initial)
    x_history.append(x_initial)
    y_history.append(sigmoid(x_initial))

    # Detener la iteración si x está cerca de 0
    if x_initial < 0.01:
        break

# Crear un rango de valores para la segunda gráfica
x_range_second = np.linspace(1, x_history[-1], len(x_history))
y_range_second = sigmoid(x_range_second)

# Graficar la función sigmoide y el proceso de gradiente descendente
plt.figure(figsize=(10, 6))

plt.plot(x_range, sigmoid_values, label='Función Sigmoidal')
plt.scatter(x_history, y_history, color='red', label='Derivada Descendente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función Sigmoidal y Derivada Descendente')
plt.grid(True)
plt.legend()

# Agregar divisiones en el eje y
y_ticks = np.arange(0, 1.2, 0.04)
plt.yticks(y_ticks)

plt.show()
```
