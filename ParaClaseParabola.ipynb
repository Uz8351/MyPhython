```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de la funcin f(x)
def f(x):
    return (x**2/8) + (x**2 - 16)

# Derivada de la funcin f(x)
def df(x):
    return (9*x/4)

# Parmetros
learning_rate = 0.01
iterations = 13000
start_x = 40

# Descenso de gradiente muy elegante.
x_values = [start_x]
for _ in range(iterations):
    gradient = df(x_values[-1])
    new_x = x_values[-1] - learning_rate * gradient
    x_values.append(new_x)
print("*********************************************")
NumeroIndice=np.argmin(x_values)
MinimoValor_x=x_values[NumeroIndice]
print("Minimo Valor x:",MinimoValor_x)
print("Minimo Valor y:",f(MinimoValor_x))
print("********************************************")
# Crear puntos para la grfica
x_vals = np.linspace(-40, 40, 100)
y_vals = f(x_vals)

# Grfica
"""
figsize=(10, 6): Este es un argumento que se pasa a
la funcin figure y establece el tamao de la figura
en pulgadas. (10, 6) significa que la figura
tendr un ancho de 10 pulgadas y una altura de 6 pulgadas.

"""
plt.figure(figsize=(14, 8))
plt.plot(x_vals, y_vals, label='f(x) = (x**2/8) + (x**2 - 6)')
plt.scatter(x_values, [f(x) for x in x_values], color='red', label='Descenso de Gradiente')
plt.scatter(start_x, f(start_x), color='green', label='Punto de inicio', marker='o')
plt.scatter(x_values[-1], f(x_values[-1]), color='black', label='Mnimo punto final', marker='o')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente')
plt.grid(True)
plt.legend()
plt.show()
```
