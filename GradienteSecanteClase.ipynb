```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci칩n secante y su derivada
def secant(x):
    return 1 / np.cos(x)

def secant_derivative(x):
    return np.tan(x) / np.cos(x)

# Par치metros de la optimizaci칩n
learning_rate = 0.01
iterations = 10
x_initial = 18

# Listas para almacenar los puntos y valores durante las iteraciones
x_values = [x_initial]
y_values = [secant(x_initial)]

# Realizar iteraciones de descenso de gradiente
for _ in range(iterations):
    x_initial = x_initial - learning_rate * secant_derivative(x_initial)
    x_values.append(x_initial)
    y_values.append(secant(x_initial))

# Crear la gr치fica
x_range = np.linspace(-15, 20, 10)
y_range = secant(x_range)

plt.figure(figsize=(15, 8))
plt.plot(x_range, y_range, label='Secante function')
plt.scatter(x_values, y_values, color='red', label='Gradient Descent steps')
plt.scatter(x_values[0], y_values[0], color='green', label='Initial point')
plt.scatter(x_values[-1], y_values[-1], color='blue', label='Final point near zero')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on Secant Function')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()
```
