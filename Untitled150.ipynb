```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# Datos de entrada (consumo de electricidad en euros)
X = np.array([[89], [97], [34], [88], [75]])
# Salida correspondiente (costo del consumo de electricidad en el mes de Agosto)
y = np.array([[100]])  # El valor real de agosto debe ser ingresado aquí

# Agregar el dato real de agosto a la lista X
X = np.append(X, y)

# Meses para el eje x del gráfico
months = ['Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto']

# Crear el perceptrón multicapa con 2 capas ocultas de 4 neuronas cada una y retropropagación
model = MLPRegressor(hidden_layer_sizes=(12, 12), max_iter=1, random_state=42, warm_start=True)

# Entrenamiento del modelo durante 1000 épocas
epochs = 1000000
for epoch in range(epochs):
    model.fit(X.reshape(-1, 1), X)

# Generar datos de los meses de marzo a agosto
X_months = np.array([[89], [97], [34], [88], [75], [0]])

# Predicción del costo para los meses de marzo a agosto
predicted_costs = model.predict(X_months)

# Redondear la predicción a dos decimales
predicted_costs = np.round(predicted_costs, 2)

# Datos reales y predicción para el gráfico de barras
real_data = X.flatten()
predicted_data = np.append(X.flatten(), predicted_costs[-1])

# Gráfico de barras con los datos reales y las predicciones para los meses de marzo a agosto
plt.bar(months, real_data, label='Datos reales', color='blue')
plt.bar('Agosto', predicted_data[-1], color='red', label=f'Predicción: {predicted_data[-1]:.2f} euros', width=0.8)
plt.xlabel('Mes')
plt.ylabel('Costo en euros')
plt.legend()
plt.show()

```
