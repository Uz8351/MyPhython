```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def func(x):
    return 2*x + (5*x)/3 + x**2

def derivative(x):
    return 2 + 5/3 + 2*x

# Parámetros de la optimización
x_initial = 40
learning_rate = 0.02
iterations = 1000

# Crear valores de x equidistantes y y correspondientes para la función
x_valores = np.linspace(-60, 60, 200)
y_valores = func(x_valores)

# Inicializar listas para almacenar valores de iteración
x_values = [x_initial]
y_values = [func(x_initial)]
print(x_values,y_values)
# Realizar iteraciones de descenso del gradiente
for _ in range(iterations):
    x_new = x_values[-1] - learning_rate * derivative(x_values[-1])
    x_values.append(x_new)
    y_values.append(func(x_new))

# Crear puntos rojos para indicar el descenso en la gráfica
red_points_x = x_values[1:-1]
red_points_y = y_values[1:-1]

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_valores, y_valores, color='black', label='Función')
plt.plot(x_values, y_values, marker='o', color='red', label='Descenso del gradiente')
#plt.scatter(red_points_x, red_points_y, color='blue', label='Puntos de descenso')

# Resaltar el punto inicial y final
plt.scatter(x_initial, func(x_initial), color='black', label='Punto Inicial')
plt.scatter(x_values[-1], y_values[-1], color='blue', label='Punto Final')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Descenso del Gradiente')
plt.grid(True)
plt.legend()
plt.show()
```
