```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def function(x):
    return (x**2 / 3) * x

def function_derivative(x):
    return  x**2

# Parámetros de la optimización
learning_rate = 0.00999991
iterations = 1000
x_initial = 18
z = 2

# Listas para almacenar los puntos y valores durante las iteraciones
x_values = [x_initial]
y_values = [function(x_initial)]

# Realizar iteraciones de descenso de gradiente
for _ in range(iterations):
    x_initial = x_initial - learning_rate * function_derivative(x_initial)
    x_values.append(x_initial)
    y_values.append(function(x_initial))

MinimoPunto=np.argmin(x_values)
MinimoValor=x_values[MinimoPunto]
print("*********************************************")
print("Valor Mínimo de x:", MinimoValor)
print("Valor Mínimo de y:",function(MinimoValor))
print("*********************************************")


# Crear la gráfica
x_range = np.linspace(-35, 35, 200)
y_range = function(x_range)

plt.figure(figsize=(12, 8))
plt.plot(x_range, y_range, label='Linea de:f(x)=(x**2 / 3) * x')
plt.scatter(x_values, y_values, color='red', label='Pasos descenso de gradiente')
plt.scatter(x_values[0], y_values[0], color='black', label='Punto inicial')
plt.scatter(x_values[-1], y_values[-1], color='blue', label='Punto más cercano a 0')

plt.xlabel('Valores x')
plt.ylabel('Valores f(x)')
plt.title('Descenso de gradiente de: f(x)=(x**2 / 3) * x')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()
```
