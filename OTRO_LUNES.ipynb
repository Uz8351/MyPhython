```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return x**2 / 3

def df(x):
    return (2/3) * x

# Parámetros iniciales
x = 18
learning_rate = 0.1
iterations = 200
x_history = [x]  # Lista para almacenar los valores de x en cada iteración
y_history = [f(x)]  # Lista para almacenar los valores de f(x) en cada iteración

# Descenso de gradiente
for _ in range(iterations):
    x = x - learning_rate * df(x)
    x_history.append(x)
    y_history.append(f(x))

# Imprimir el valor mínimo encontrado
print("Valor mínimo encontrado:", x)

# Crear un rango de valores para la gráfica
x_range = np.linspace(0, max(x_history), 100)
y_range = f(x_range)

# Graficar la parábola, la línea de regresión y los puntos de descenso
plt.plot(x_range, y_range, label='Parábola $f(x) = x^2/3$')
plt.scatter(x_history, y_history, color='black', label='Puntos de descenso')
plt.scatter(x, f(x), color='red', label='Punto mínimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Agregar divisiones en el eje y
y_ticks = np.arange(0, 150, 10)
plt.yticks(y_ticks)

plt.show()
```
