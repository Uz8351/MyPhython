```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from PIL import Image

# Cargar la imagen
image_path = '/content/drive/MyDrive/aguila.jpg'
img = Image.open(image_path)

# Preprocesamiento de la imagen
img = img.resize((224, 224))  # Ajusta el tamaño según tus necesidades
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)

# Definir el modelo
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(50, activation='relu'))
model.add(layers.Dense(20, activation='relu'))
model.add(layers.Dense(20, activation='relu'))
model.add(layers.Dense(6, activation='softmax'))  # Capa de salida con 6 neuronas

# Compilar el modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Resumen del modelo
model.summary()

# Visualizar la imagen
plt.imshow(img)
plt.title('Imagen Original')
plt.show()

# Visualizar la imagen después de la capa convolucional
conv_layer = models.Model(inputs=model.inputs, outputs=model.layers[0].output)
feature_maps = conv_layer(img_array)
plt.figure(figsize=(8, 8))
for i in range(32):
    plt.subplot(4, 8, i+1)
    plt.imshow(feature_maps[0, :, :, i], cmap='viridis')
    plt.axis('off')
plt.suptitle('Activaciones de la Capa Convolucional')
plt.show()


import tensorflow as tf
from tensorflow.keras import layers, models

# Definir el modelo hasta la capa de aplanamiento
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())

# Imprimir el resumen del modelo para obtener la forma después de la aplanación
model.summary()
import tensorflow as tf
from tensorflow.keras import layers, models

# Definir el modelo hasta la capa de aplanamiento
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())

# Obtener la forma después de la aplanación
output_shape = model.output_shape[1]
print("Output Shape después de la aplanación:", output_shape)
```
