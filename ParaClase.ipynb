```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return x**2

def df(x):
    return 2 * x

# Parámetros
x=x_start = 18
learning_rate = 0.1
iterations = 10

# Inicializar listas para almacenar los valores de x, y y
x_values = [x_start]
y_values = [f(x_start)]

# Descenso de gradiente
#x = x_start
for _ in range(iterations):
    x_values.append(x)

    y_values.append(f(x))

    x -= learning_rate * df(x)

# Graficar la función y el descenso de gradiente
x_vals = np.linspace(-20, 20, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2')
plt.scatter(x_start, f(x_start), color='green', label='Punto de partida')
plt.scatter(x_values[-1], y_values[-1], color='red', label='Punto final')
plt.scatter(x_values, y_values, color='red', label='Descenso de gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Descenso de Gradiente en f(x) = x^2')
plt.grid(True)
plt.legend()
plt.show()
```
