```python
import numpy as np
import matplotlib.pyplot as plt

# Nuevos datos de ejemplo (puntos en forma de coordenadas x, y)
x = np.array([1, 2, 3, 4, 5, 3, 5, 4, 5, 6, 7, 8, 2, 3, 4, 5, 3, 4, 5, 6, 6, 7,3,4,3,2,3,6,7,8,6,6,5,4,5,1])
y = np.array([2, 4, 5, 4, 5, 12, 3, 4, 7, 8, 5, 6, 7, 8, 3, 2, 2, 4, 5, 6, 6, 1,4,6,7,8,7,4,5,6,3,2,5,7,8,1])

# Inicialización de parámetros (pendiente y ordenada al origen)
slope = np.random.rand()
intercept = np.random.rand()
learning_rate = 0.01
iterations = 20

# Descenso de gradiente
for i in range(iterations):
    predictions = slope * x + intercept
    errors = predictions - y
    gradient_slope = (2 / len(x)) * np.sum(errors * x)
    gradient_intercept = (2 / len(x)) * np.sum(errors)
    slope -= learning_rate * gradient_slope

    intercept -= learning_rate * gradient_intercept

# Generar valores de x para la gráfica
x_vals = np.linspace(min(x), max(x), num=100)

# Configuración de tamaño de figura (más grande)
plt.figure(figsize=(20.0/2.54 ,12.0/2.54))

# Graficar los puntos y la línea ajustada
plt.scatter(x=x, y=y, label='Datos')
plt.plot(x_vals, slope*x_vals + intercept, color='red', label='Ajuste')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print(f"Pendiente final: {slope}")
print(f"Ordenada al origen final: {intercept}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo a minimizar
def f(x):
    return x**2 + 2*x + 1

# Gradiente de la función objetivo
def grad_f(x):
    return 2*x + 2

# Datos de entrenamiento
X = np.array([1, 2, 3, 4, 5])
y = f(X)

# Inicializamos los parámetros
w = np.array([0, 0])

# Variables para guardar la información de la optimización
history = []

# Iteramos sobre los datos de entrenamiento
for x, y_true in zip(X, y):
    # Calculamos el gradiente
    g = grad_f(x)

    # Actualizamos los parámetros
    w = w - 0.01 * g

    # Guardamos información para la gráfica
    history.append((w[0], w[1], f(x)))

# Crear la gráfica
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Extraer datos para la gráfica
w0_vals = [h[0] for h in history]
w1_vals = [h[1] for h in history]
f_vals = [h[2] for h in history]

# Graficar la optimización en 3D
ax.plot(w0_vals, w1_vals, f_vals, marker='o')

ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Función Objetivo')

plt.show()

print("Parámetros finales:", w)
```
