```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de entrada sintéticos
np.random.seed(0)
X = np.random.rand(100, 6)

# Datos de prueba
X_test = np.random.rand(10, 6)  # 10 ejemplos de prueba con las mismas características

# Etiquetas reales para los datos de prueba
y_test = np.random.rand(10, 1)  # Reemplaza esto con tus etiquetas reales

# Definir pesos y sesgos
input_size = 6
hidden_size = 8
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Función de activación
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Propagación hacia adelante
z1 = np.dot(X, W1) + b1
a1 = sigmoid(z1)

z2 = np.dot(a1, W2) + b2
a2 = sigmoid(z2)

# Propagación hacia adelante para los datos de prueba
z1_test = np.dot(X_test, W1) + b1
a1_test = sigmoid(z1_test)

z2_test = np.dot(a1_test, W2) + b2
a2_test = sigmoid(z2_test)

# Cálculo del error cuadrático medio (MSE) para evaluar el rendimiento
mse = np.mean((a2_test - y_test)**2)
print(f'Error cuadrático medio (MSE) en los datos de prueba: {mse:.3f}')

# Gráfico de las salidas
plt.figure(figsize=(12, 5))
plt.scatter(range(100), a2, label='Salida de la Red Neuronal (Entrenamiento)')
plt.scatter(range(100, 110), a2_test, label='Salida de la Red Neuronal (Prueba)')

# Resalta los puntos con un MSE mayor que el promedio
mean_mse = np.mean((a2_test - y_test)**2)
for i in range(len(a2_test)):
    mse_i = (a2_test[i] - y_test[i])**2
    if mse_i > mean_mse:
        plt.annotate(f'MSE: {mse_i[0]:.3f}', (i + 100, a2_test[i]), fontsize=8, color='red')

plt.xlabel('Ejemplo')
plt.ylabel('Salida')
plt.legend()
plt.show()
```
