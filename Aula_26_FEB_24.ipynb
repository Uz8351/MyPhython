```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential

# Cargar la imagen
ruta_imagen = "/content/drive/MyDrive/druida/flores/flores1/image_00004.jpg"
imagen_original = cv2.imread(ruta_imagen)

kernel_random = np.random.randn(2, 3)
# Crear un filtro de convolucin (kernel)
kernel = np.random.randint(1, 4, size=(3, 3))/np.sum(kernel_random)

# Aplicar la convolucin
imagen_convolucionada = cv2.filter2D(imagen_original, -1, kernel)

# Crear el modelo CNN
model = Sequential()
model.add(Conv2D(3, (3, 3), activation='relu', input_shape=(imagen_convolucionada.shape[0], imagen_convolucionada.shape[1], 3)))
model.add(MaxPooling2D(pool_size=(3, 3)))

# Preparar la imagen para la entrada del modelo CNN
imagen_convolucionada = imagen_convolucionada.reshape((1, imagen_convolucionada.shape[0], imagen_convolucionada.shape[1], 3))

# Aplicar la convolucin y pooling utilizando el modelo CNN
imagen_resultante = model.predict(imagen_convolucionada)

# Visualizar la imagen resultante
imagen_resultante = np.squeeze(imagen_resultante, axis=0)
plt.imshow(imagen_resultante.astype('uint8'))
plt.title("Imagen procesada con convolucin y pooling")
plt.axis("off")
plt.show()
```
