```python
"""
Función de partida para los valores de
Entrada y de Salida.
f(x,y,t)=(x**2/6)+(y*t)
[1,6,8]
36,5

"""
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
"""
Debe instalarse, si no está instalado, seaborn.
La forma de instalarno en Ubuntu-Linux es:
pip install seaborn
"""

Valores_Entrada = np.array([[1, 2, 4], [4, 3, 5], [2, 9, 6], [7, 3, 4], [8, 1, 1],
                            [2, 4, 8], [7, 5, 2], [4, 1, 6], [5, 9, 3], [4, 5, 7],
                            [3, 7, 5], [21, 2, 1], [12, 1, 2], [2, 17,2],[2,3,12]])

Valores_Salida = np.array([8.166, 17.666, 54.666, 20.166, 11.666,
                           32.666, 18.166, 8.666, 31.666, 37.666,
                           36.5, 75.5, 26, 34.166,36.666])

print("Forma de Valores_Entrada:", Valores_Entrada.shape)
print("Forma de Valores_Salida:", Valores_Salida.shape)


#Vamos a diseñar nuestra red neuronal.
Nuestra_input=tf.keras.Input(shape=(Valores_Entrada.shape[1]))
#Neuro_Intermedia=tf.keras.layers.Dense(128, activation='relu')(Nuestra_input)
#Aquí hay un ejemplo de cómo podrías incluir una inicialización
#de pesos en la capa Neuro_Intermedia utilizando la inicialización Xavier (Glorot):
Neuro_Intermedia = tf.keras.layers.Dense(128, activation='relu', kernel_initializer='glorot_uniform')(Nuestra_input)
Neuro_Intermedia=tf.keras.layers.Dense(128, activation='relu')(Neuro_Intermedia)
#Neuro_Intermedia=tf.keras.layers.Dense(128, activation='relu')(Neuro_Intermedia)
Neurona_Output=tf.keras.layers.Dense(1)(Neuro_Intermedia)

#Procedemos a establecer el modelo.
El_Modelo=tf.keras.Model(Nuestra_input,Neurona_Output)

#Compilamos el modelo con la funcion de pérdida adecuada para la regresión.
El_Modelo.compile(optimizer='adam', loss='mean_squared_error')

#Ahora ya podemos entrenar el modelo.
El_Modelo.fit(Valores_Entrada,Valores_Salida,epochs=20000,verbose=0)
#Eligimos un valor de entrada para hacer una predicción, por ejemplo: [3,7,5]
Entrada_Nueva=np.array([[1,6,8]])

Nuestra_Prediccion=El_Modelo.predict(Entrada_Nueva)

print("Valor Real de los valores [1,6,8]=48,166 El valor de la predicción es el siguiente:",Nuestra_Prediccion)
plt.figure(figsize=(10,6))
plt.scatter(Valores_Entrada[:,0],Valores_Salida, color='red', label='Valores de Entrada')
plt.scatter(Entrada_Nueva[:,0],Nuestra_Prediccion,color='blue',label='Nuestra_Prediccion')
#sns.regplot(x=Valores_Entrada[:0],y=Valores_Salida, color='red',scatter=False, label='Linea de regresion')
#sns.regplot(x=Valores_Entrada[:0], y=Valores_Salida, color='red', scatter=False, label='Linea de regresion')
sns.regplot(x=Valores_Entrada[:, 0], y=Valores_Salida, color='red', scatter=False, label='Linea de regresion')
plt.xlabel('Valores en el eje x')
plt.ylabel('Valores en el eje y')
plt.legend()
plt.show()
```
