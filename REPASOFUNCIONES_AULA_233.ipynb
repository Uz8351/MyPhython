```python

import matplotlib.pyplot as plt

def MiFuncion(ConstanteAbierta, ConstanteCerrada, FactorRiesgo):
	mayoracion=ConstanteAbierta*ConstanteCerrada/FactorRiesgo
	gradiente=ConstanteAbierta/ConstanteCerrada
	media=ConstanteAbierta+ConstanteCerrada/2
	factorfinal=FactorRiesgo/media
	return mayoracion, gradiente, media, factorfinal

print("Introduce valores:")


ConstanteAbierta = float(input("Introduce ConstanteAbierta:"))
ConstanteCerrada=float(input("Introduce ConstanteCerrada:"))
FactorRiesgo=float(input("Introduce FactorRiesgo:"))
mayoracion,gradiente,media,factorfinal=MiFuncion(ConstanteAbierta, ConstanteCerrada, FactorRiesgo)

print("mayoracion:",mayoracion)
print("gradiente",gradiente)
print("media:",media)
print("factorial:",media)

# Crear un gráfico de barras con los resultados
nombres = ['Mayoración', 'Gradiente', 'Media', 'Factor Final']
valores = [mayoracion, gradiente, media, factorfinal]

plt.bar(nombres, valores, color='red')
plt.title("Resultados de la función MiFuncion")
plt.xlabel("Variables")
plt.ylabel("Valores")
plt.show()
```
