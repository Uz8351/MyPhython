```python
"""
f(x)=x**2/3+x**2
Derivada=8*x/3

"""
import numpy as np
import matplotlib.pyplot as plt
def Funcion(x):
	return x**2/3+x**2

def FuncionDerivada(x):
	return 8*x/3

Valores_x=[]
Valores_y=[]
RatioAprendizaje=0.01
Iteraciones=1000
PuntoInicial_x=8

for _ in range(Iteraciones):
    x_Inicial = [PuntoInicial_x]
    y_Inicial = Funcion(PuntoInicial_x)
    Valores_x.append(PuntoInicial_x)
    Valores_y.append(y_Inicial)
    PuntoInicial_x -= FuncionDerivada(PuntoInicial_x) * RatioAprendizaje
# Encontrar el índice del valor mínimo de y
"""
indice_minimo_y = np.argmin(Valores_y)
Esta línea utiliza la función np.argmin()
de la librería NumPy para encontrar el índice
del valor mínimo en un array
(en este caso, el array Valores_y que contiene
los valores de la función evaluada en diferentes puntos).
"""

# Encontrar el índice del último punto de descenso
indice_ultimo_descenso = -1

# Imprimir los valores mínimos de x e y
print("***************************************************")
print("Valor mínimo de x:", Valores_x[indice_minimo_y])
print("Valor mínimo de y:", Valores_y[indice_minimo_y])
print("***************************************************")

RangoDe_x = np.linspace(-10, 10, 100)
RangoDe_y = Funcion(RangoDe_x)

plt.plot(RangoDe_x, RangoDe_y, color='black', label='Parábola  f(x)=x**2/3+x**2')
plt.scatter(Valores_x, Valores_y, color='red', label='Puntos de descenso')
plt.scatter(Valores_x[0], Valores_y[0], color='green', label='Punto inicial')  # Punto inicial en verde
plt.scatter(Valores_x[indice_ultimo_descenso], Valores_y[indice_ultimo_descenso], color='blue', label='Último punto')  # Último punto en azul

plt.xlabel('Eje x')
plt.ylabel('Eje y, valores f(x)=x**2/3+x**2')
plt.title('Descenso de Gradiente en una Parábola')
plt.grid(True)
plt.legend()
plt.show()
print(Valores_x)
```
