```python
"""
En este código, primero solicitamos al usuario
el tamaño de la matriz (número de filas y columnas).
Luego, creamos una matriz vacía con el tamaño especificado
y solicitamos al usuario ingresar los valores
para cada celda de la matriz.

Después de almacenar los datos en la matriz,
la imprimimos en la consola. Luego,
creamos una figura y un subgráfico
utilizando Matplotlib.
Utilizamos la función imshow()
para mostrar los datos de la matriz en una gráfica,
y configuramos los ejes y la apariencia de la gráfica.
"""
#Importamos los modulos necesarios
#numpy para arrays, y  matplotlib para los gráficos
import numpy as nphim
import matplotlib.pyplot as plthim

# Solicitamos al usuario el tamaño de la matriz
#que queremos utilizar
filas = int(input("Ingrese el número de filas: "))
columnas = int(input("Ingrese el número de columnas: "))

# Ahora creamos una matriz vacía con el tamaño especificado
matriz = nphim.empty((filas, columnas))

# Solicitamos al usuario ingresar los datos de la matriz
for i in range(filas):
    for j in range(columnas):
        valor = float(input(f"Ingrese el valor para la posición ({i}, {j}): "))
        matriz[i, j] = valor

# Imprimimos la matriz
print("La matriz ingresada es:")
print(matriz)

# Creaamos una figura y un subgráfico para la gráfica
fig, ax = plthim.subplots()

# Graficamos los datos de la matriz
ax.imshow(matriz, cmap='viridis')

# Configuramos los ejes de la gráfica
ax.set_xticks(nphim.arange(columnas))
ax.set_yticks(nphim.arange(filas))
ax.set_xticklabels(nphim.arange(columnas))
ax.set_yticklabels(nphim.arange(filas))
ax.tick_params(axis='both', which='both', length=0)

# Mostramos los valores en cada celda de la matriz
for i in range(filas):
    for j in range(columnas):
        ax.text(j, i, f"{matriz[i, j]:.2f}", ha='center', va='center', color='w')

# Configuramos la apariencia de la gráfica
plthim.title("Datos de la Matriz")
plthim.xlabel("Columnas")
plthim.ylabel("Filas")
plthim.grid(visible=False)

# Mostramos la gráfica
plthim.show()
```
