```python
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def f(x):
    return x**2

# Gradiente de la función objetivo
def grad_f(x):
    return 2*x

# Valores iniciales
x_0 = 1.0

# Parámetros de Adam
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8

# Inicialización de los parámetros de Adam
m = 0
v = 0
t = 0

# Listas para almacenar los valores de x y f(x) en cada iteración
x_values = [x_0]
f_values = [f(x_0)]

# Iteraciones
for i in range(100):
    t += 1
    gradient = grad_f(x_0)
    m = beta1 * m + (1 - beta1) * gradient
    v = beta2 * v + (1 - beta2) * (gradient**2)
    m_hat = m / (1 - beta1**t)
    v_hat = v / (1 - beta2**t)
    step_size = m_hat / (np.sqrt(v_hat) + epsilon)
    x_0 -= step_size

    # Almacenar los valores de x y f(x) en cada iteración
    x_values.append(x_0)
    f_values.append(f(x_0))

# Gráfica del descenso de gradiente
plt.plot(range(101), f_values)  # Cambié 100 a 101 para incluir la iteración inicial
plt.xlabel("Iteraciones")
plt.ylabel("f(x)")
plt.show()
```
