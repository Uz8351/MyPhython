```python
import numpy as np
import matplotlib.pyplot as plt

# Función a optimizar
def f(x, y):
    return (7 * x * y) / np.exp(x**2 + y**2)

# Gradiente de la función
def gradient(x, y):
    df_dx = (7 * y * (2 * x * np.exp(x**2 + y**2) - x)) / np.exp(x**2 + y**2)**2
    df_dy = (7 * x * (2 * y * np.exp(x**2 + y**2) - y)) / np.exp(x**2 + y**2)**2
    return df_dx, df_dy

# Inicialización
learning_rate = 0.1
iterations = 50
x, y = 1, 1

# Listas para almacenar valores de x, y y costos
x_values = [x]
y_values = [y]
costs = [f(x, y)]

# Descenso de gradiente
for i in range(iterations):
    df_dx, df_dy = gradient(x, y)
    x -= learning_rate * df_dx
    y -= learning_rate * df_dy
    x_values.append(x)
    y_values.append(y)
    costs.append(f(x, y))

# Graficar el descenso de gradiente
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_values, y_values, c='r', marker='o')
plt.plot(x_values, y_values, 'b-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Descenso de Gradiente')

# Graficar la función de costo
plt.subplot(1, 2, 2)
plt.plot(range(iterations+1), costs, 'g-')
plt.xlabel('Iteraciones')
plt.ylabel('Costo')
plt.title('Función de Costo')

plt.tight_layout()
plt.show()
```
